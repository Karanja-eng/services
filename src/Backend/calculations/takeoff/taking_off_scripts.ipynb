```python
# septic_takeoff_app.py
def calculate_septic_takeoff():
    print("Septic Tank Quantity Takeoff Calculator")
    
    # Inputs
    tank_int_L = float(input("Tank internal length (m): "))
    tank_int_W = float(input("Tank internal width (m): "))
    manhole_int_L = float(input("Manhole internal length (m): "))
    manhole_int_W = float(input("Manhole internal width (m): "))
    wall_thick = float(input("Wall thickness (m): "))
    bed_thick_tank = float(input("Tank bed thickness (m): "))
    bed_thick_manhole = float(input("Manhole bed thickness (m): "))
    blinding_thick = float(input("Blinding thickness (m): "))
    slab_thick = float(input("Slab thickness (m): "))
    tank_depth = float(input("Tank total depth from ground (m): "))
    manhole_depth = float(input("Manhole depth from ground (m): "))
    veg_soil = float(input("Vegetable soil to remove (m): "))
    working_space = float(input("Working space allowance each side (m): "))
    cover_size_L = float(input("Cover length (m): "))
    cover_size_W = float(input("Cover width (m): "))
    num_covers = int(input("Number of covers: "))
    
    # Excavation dims
    excav_L_tank = tank_int_L + 2 * (wall_thick + working_space)
    excav_W_tank = tank_int_W + 2 * (wall_thick + working_space)
    excav_L_manhole = manhole_int_L + 2 * (wall_thick + working_space)
    excav_W_manhole = manhole_int_W + 2 * (wall_thick + working_space)
    
    # Site Clearance
    clear_tank = excav_L_tank * excav_W_tank
    clear_manhole = excav_L_manhole * excav_W_manhole
    clearance = clear_tank + clear_manhole
    
    # Excavation
    avg_depth_tank = tank_depth + blinding_thick
    excav_tank = excav_L_tank * excav_W_tank * avg_depth_tank
    excav_manhole = excav_L_manhole * excav_W_manhole * (manhole_depth + veg_soil)
    excavation = excav_tank + excav_manhole
    
    # Blinding
    blinding = excav_L_tank * excav_W_tank
    
    # Concrete Bed
    bed_tank = excav_L_tank * excav_W_tank * bed_thick_tank
    bed_manhole = manhole_int_L * manhole_int_W * bed_thick_manhole
    concrete_bed = bed_tank + bed_manhole
    
    # Walls
    girth_tank = 2 * (tank_int_L + tank_int_W)
    height_tank = tank_depth - slab_thick
    wall_tank = girth_tank * height_tank * wall_thick
    
    girth_manhole = 2 * (manhole_int_L + manhole_int_W) - wall_thick
    height_manhole = manhole_depth - slab_thick
    wall_manhole = girth_manhole * height_manhole * wall_thick
    walls = wall_tank + wall_manhole
    
    # Slab
    slab_area_tank = (tank_int_L + 2*wall_thick) * (tank_int_W + 2*wall_thick)
    slab_area_manhole = manhole_int_L * manhole_int_W
    openings = num_covers * cover_size_L * cover_size_W
    slab_net = slab_area_tank + slab_area_manhole - openings
    slab_conc = slab_net * slab_thick
    
    # Formwork (soffit + edges)
    soffit_tank = tank_int_L * tank_int_W
    soffit_manhole = manhole_int_L * manhole_int_W
    edges = 2*(tank_int_L + tank_int_W) + 2*(manhole_int_L + manhole_int_W) + num_covers * 2*(cover_size_L + cover_size_W)
    formwork = soffit_tank + soffit_manhole + edges * slab_thick
    
    # Backfill
    backfill_depth_tank = height_tank
    backfill_depth_manhole = height_manhole
    backfill_tank = girth_tank * backfill_depth_tank * wall_thick
    backfill_manhole = girth_manhole * backfill_depth_manhole * wall_thick
    backfill = backfill_tank + backfill_manhole
    
    # Output
    print("\n--- QUANTITIES ---")
    print(f"Site Clearance: {clearance:.2f} m²")
    print(f"Excavation: {excavation:.2f} m³")
    print(f"Blinding (75mm): {blinding:.2f} m²")
    print(f"Concrete Bed: {concrete_bed:.2f} m³")
    print(f"Wall Concrete: {walls:.2f} m³")
    print(f"Slab Concrete: {slab_conc:.2f} m³")
    print(f"Formwork: {formwork:.2f} m²")
    print(f"Backfill: {backfill:.2f} m³")
    
    return {
        'clearance_m2': clearance,
        'excavation_m3': excavation,
        'blinding_m2': blinding,
        'bed_m3': concrete_bed,
        'walls_m3': walls,
        'slab_m3': slab_conc,
        'formwork_m2': formwork,
        'backfill_m3': backfill
    }

# Run
if __name__ == "__main__":
    quantities = calculate_septic_takeoff()
```

```python
# staircase_takeoff_app.py
import math

def calculate_stair_takeoff():
    print("Staircase Quantity Takeoff Calculator (Dog-leg with 2 Flights)")

    # Inputs
    clear_width = float(input("Clear stair width (m): "))
    wall_thick = float(input("Wall thickness (m): "))
    waist_thick = float(input("Waist/slab thickness (m): "))
    landing_L = float(input("Landing length (m): "))
    landing_W = float(input("Landing width (m): "))
    tread = float(input("Tread/going (m): "))
    rise = float(input("Riser height (m): "))
    num_risers_f1 = int(input("Number of risers in flight 1: "))
    num_treads_f1 = int(input("Number of treads in flight 1 (may differ from risers): "))
    num_risers_f2 = int(input("Number of risers in flight 2: "))
    num_treads_f2 = int(input("Number of treads in flight 2: "))
    rebar_spacing = float(input("Main rebar spacing (m): "))
    rebar_bend = float(input("Rebar bend/anchor length each end (m): "))
    bal_spacing = float(input("Baluster spacing (m): "))
    bal_height = float(input("Balustrade height (m): "))

    # Calculations
    horiz_f1 = num_treads_f1 * tread
    vert_f1 = num_risers_f1 * rise
    hypo_f1 = math.sqrt(horiz_f1**2 + vert_f1**2)

    horiz_f2 = num_treads_f2 * tread
    vert_f2 = num_risers_f2 * rise
    hypo_f2 = math.sqrt(horiz_f2**2 + vert_f2**2)

    total_risers = num_risers_f1 + num_risers_f2

    conc_width = clear_width + wall_thick / 2
    form_width = clear_width

    form_landing_w = landing_W - tread if tread < landing_W else landing_W
    step_hypo = math.sqrt(tread**2 + rise**2)

    # Formwork
    form_soffit_landing = landing_L * form_landing_w
    form_soffit_f1 = hypo_f1 * form_width
    form_soffit_f2 = hypo_f2 * form_width
    form_soffit_total = form_soffit_landing + form_soffit_f1 + form_soffit_f2

    form_riser_edges = total_risers * clear_width  # linear m
    form_string_lm = hypo_f1 + hypo_f2  # linear m for edges

    # Concrete
    conc_landing = landing_L * landing_W * waist_thick
    conc_waist_f1 = hypo_f1 * conc_width * waist_thick
    conc_waist_f2 = hypo_f2 * conc_width * waist_thick
    conc_steps = total_risers * (0.5 * tread * rise) * conc_width
    conc_total = conc_landing + conc_waist_f1 + conc_waist_f2 + conc_steps

    # Reinforcement
    num_main_bars = int(conc_width / rebar_spacing) + 1
    rebar_len_landing = landing_L + 2 * rebar_bend
    rebar_len_f1 = hypo_f1 + 2 * rebar_bend
    rebar_len_f2 = hypo_f2 + 2 * rebar_bend
    rebar_main_total = num_main_bars * (rebar_len_landing + rebar_len_f1 + rebar_len_f2)

    dist_bars_per_step = 2  # assume 2 per step
    dist_bar_len = conc_width
    dist_bars_total = total_risers * dist_bars_per_step * dist_bar_len

    # Balustrade
    rail_len = hypo_f1 + hypo_f2 + landing_W  # approximate total rail length
    num_balusters = int(rail_len / bal_spacing)
    bal_total_len = num_balusters * bal_height

    num_standards = num_balusters // 2 + 2  # approximate

    # Finishes
    terrazzo_landing = landing_L * landing_W
    terrazzo_treads = total_risers * tread * clear_width
    terrazzo_risers = total_risers * rise * clear_width
    terrazzo_total = terrazzo_landing + terrazzo_treads + terrazzo_risers

    plaster_soffit = form_soffit_total

    nonslip_lm = total_risers * clear_width

    # Output
    print("\n--- QUANTITIES ---")
    print(f"Formwork Soffit Total: {form_soffit_total:.2f} m²")
    print(f"Formwork Riser Edges: {form_riser_edges:.2f} lm")
    print(f"Formwork String Edges: {form_string_lm:.2f} lm")
    print(f"Concrete Total: {conc_total:.2f} m³")
    print(f"Main Rebar Total: {rebar_main_total:.2f} m")
    print(f"Distribution Bars Total: {dist_bars_total:.2f} m")
    print(f"Balustrade Rail Length: {rail_len:.2f} m")
    print(f"Balusters Total Length: {bal_total_len:.2f} m")
    print(f"Number of Standards (approx): {num_standards}")
    print(f"Terrazzo Total Area: {terrazzo_total:.2f} m²")
    print(f"Plaster Soffit Area: {plaster_soffit:.2f} m²")
    print(f"Non-slip Inserts: {nonslip_lm:.2f} lm")

    return {
        'form_soffit_m2': form_soffit_total,
        'form_riser_lm': form_riser_edges,
        'form_string_lm': form_string_lm,
        'conc_m3': conc_total,
        'rebar_main_m': rebar_main_total,
        'rebar_dist_m': dist_bars_total,
        'rail_m': rail_len,
        'balusters_m': bal_total_len,
        'terrazzo_m2': terrazzo_total,
        'plaster_m2': plaster_soffit,
        'nonslip_lm': nonslip_lm
    }

# Run
if __name__ == "__main__":
    quantities = calculate_stair_takeoff()
```

```python
# roadworks_takeoff_professional.py
import math
from typing import Dict

def calculate_roadworks_takeoff():
    print("Professional Roadworks Quantity Takeoff (CESMM4 Compliant)")
    print("Based on video analyses: Site clearance, earthworks, pavement layers, drainage, landscaping.")
    
    # Inputs from drawings (in m)
    site_length = float(input("Site length (m): "))
    site_width = float(input("Site width (m): "))
    house_length = float(input("House length (m): "))
    house_width = float(input("House width (m): "))
    parking_length = float(input("Parking length (m): "))
    parking_width = float(input("Parking width (m): "))
    driveway_length = float(input("Driveway length (m): "))
    driveway_width = float(input("Driveway width (m): "))
    bellmouth_r1 = float(input("Bellmouth radius 1 (m): "))
    bellmouth_r2 = float(input("Bellmouth radius 2 (m): "))
    veg_depth = float(input("Vegetable soil depth (m): "))
    excav_depth = float(input("Excavation depth after veg soil (m): "))
    murram_depth_avg = float(input("Average murram filling depth (m): "))
    hardcore_thick = float(input("Hardcore thickness (m): "))
    bitumen_thick = float(input("Bitumen premix thickness (m): "))
    topsoil_thick = float(input("Imported topsoil thickness (m): "))
    kerb_length = float(input("Kerb/channel length straight (m): "))
    fence_length1 = float(input("Fence length type 1 (m): "))
    fence_length2 = float(input("Fence length type 2 (m): "))
    num_trees_small = int(input("Number of small trees (G 0.5-2m): "))
    num_trees_large = int(input("Number of large trees (G >2m): "))
    num_stumps = int(input("Number of stumps (<1m): "))
    building_vol = float(input("Building demolition volume (m³): "))
    pipe_length = float(input("Pipeline removal length (m): "))
    num_mahogany = int(input("Number of mahogany trees: "))
    hedge_length = float(input("Hedge length (m): "))
    num_gates = int(input("Number of gates: "))
    backing_allow = float(input("Backing allowance each side (m): "))
    
    # Site Clearance (Class D)
    clear_area = (site_length * site_width) - (house_length * house_width)
    
    # Trees and Stumps (assume diameters for G, but count-based)
    tree_g1 = num_trees_small  # nos
    tree_g2 = num_trees_large  # nos
    stumps = num_stumps  # nos
    
    # Building and Pipeline
    build_demol = building_vol  # m³
    pipe_rem = pipe_length  # m
    
    # Earthworks (Class E)
    # Veg Soil Excavation
    veg_area = clear_area  # m²
    veg_vol = veg_area * veg_depth  # m³
    
    # Mass Excavation
    park_l_adj = parking_length + 2 * backing_allow
    park_w_adj = parking_width + 2 * backing_allow
    park_exc_vol = park_l_adj * park_w_adj * excav_depth  # m³
    
    drive_l_adj = driveway_length + 2 * backing_allow
    drive_w_adj = driveway_width + 2 * backing_allow
    drive_exc_vol = drive_l_adj * drive_w_adj * excav_depth  # m³
    
    # Bellmouth (sector approx as (3/14)*pi*r^2 per video)
    bell_area1 = (3/14) * math.pi * bellmouth_r1**2
    bell_area2 = (3/14) * math.pi * bellmouth_r2**2
    bell_area = bell_area1 + bell_area2  # m²
    bell_exc_vol = bell_area * excav_depth  # m³
    
    total_excav = park_exc_vol + drive_exc_vol + bell_exc_vol  # m³
    total_excav_veg = veg_vol  # m³ (separate)
    
    # Disposal (assume same as excavated vols)
    disposal_veg = veg_vol  # m³
    disposal_other = total_excav  # m³
    
    # Preparation Surfaces (areas)
    prep_area = (park_l_adj * park_w_adj) + (drive_l_adj * drive_w_adj) + bell_area  # m²
    
    # Filling (Class E)
    # Murram
    park_area = park_l_adj * park_w_adj  # m²
    drive_area = drive_l_adj * drive_w_adj  # m²
    murram_vol_park = park_area * murram_depth_avg
    murram_vol_drive = drive_area * murram_depth_avg
    murram_vol_bell = bell_area * murram_depth_avg
    total_murram = murram_vol_park + murram_vol_drive + murram_vol_bell  # m³
    
    # Hardcore (area-based per video)
    hardcore_area = park_area + drive_area + bell_area  # m²
    hardcore_vol = hardcore_area * hardcore_thick  # m³ (though measured m², vol for consistency)
    
    # Imported Topsoil for Grass
    grass_area = site_width * (site_length - driveway_width - parking_width)  # approx m²
    topsoil_vol = grass_area * topsoil_thick  # m³
    
    # Trim/Prep Filled Surfaces
    trim_topsoil = grass_area  # m²
    prep_filled_other = hardcore_area  # m²
    
    # Pavement (Class R)
    # Bitumen Premix
    bitumen_area = park_area + drive_area + bell_area  # m²
    bitumen_vol = bitumen_area * bitumen_thick  # m³
    
    # PCC Kerb/Channel Straight
    kerb_straight = kerb_length  # m
    
    # Curved Kerb (approx bell circumference quarter)
    kerb_curved = (math.pi * (bellmouth_r1 + bellmouth_r2)) / 4  # m approx
    
    # Landscaping (Class E)
    mahogany_nos = num_mahogany  # nos
    hedge_len_net = hedge_length - (backing_allow * 2)  # m approx
    grass_seed_area = grass_area  # m²
    
    # Miscellaneous (Class X)
    fence1_len = fence_length1  # m
    fence2_len = fence_length2  # m
    gates_nos = num_gates  # nos
    
    # Output
    print("\n--- QUANTITIES (Class D: Demolition & Clearance) ---")
    print(f"Site Clearance: {clear_area:.2f} m²")
    print(f"Trees (0.5-2m G): {tree_g1} nos")
    print(f"Trees (>2m G): {tree_g2} nos")
    print(f"Stumps (<1m): {stumps} nos")
    print(f"Building Demolition: {build_demol:.2f} m³")
    print(f"Pipeline Removal (225mm): {pipe_rem:.2f} m")
    
    print("\n--- QUANTITIES (Class E: Earthworks) ---")
    print(f"Veg Soil Excavation: {veg_vol:.2f} m³")
    print(f"Mass Excavation (other): {total_excav:.2f} m³")
    print(f"Disposal Veg Soil: {disposal_veg:.2f} m³")
    print(f"Disposal Other Material: {disposal_other:.2f} m³")
    print(f"Prep Excavated Surfaces: {prep_area:.2f} m²")
    print(f"Murram Filling: {total_murram:.2f} m³")
    print(f"Hardcore Filling: {hardcore_area:.2f} m² ({hardcore_vol:.2f} m³)")
    print(f"Imported Topsoil: {topsoil_vol:.2f} m³")
    print(f"Trim Filled Topsoil: {trim_topsoil:.2f} m²")
    print(f"Prep Filled Other: {prep_filled_other:.2f} m²")
    
    print("\n--- QUANTITIES (Class R: Roads & Pavings) ---")
    print(f"Bitumen Premix: {bitumen_area:.2f} m² ({bitumen_vol:.2f} m³)")
    print(f"PCC Kerb/Channel Straight: {kerb_straight:.2f} m")
    print(f"PCC Kerb/Channel Curved: {kerb_curved:.2f} m")
    
    print("\n--- QUANTITIES (Class E: Landscaping) ---")
    print(f"Grass Seeding: {grass_seed_area:.2f} m²")
    print(f"Mahogany Trees (1m high): {mahogany_nos} nos")
    print(f"Euphorbia Hedge (0.5m high): {hedge_len_net:.2f} m")
    
    print("\n--- QUANTITIES (Class X: Miscellaneous) ---")
    print(f"Fence Type 1 (2-2.5m): {fence1_len:.2f} m")
    print(f"Fence Type 2 (1.5-2m): {fence2_len:.2f} m")
    print(f"Metal Gates (>5m): {gates_nos} nos")
    
    return {
        # Class D
        'clear_m2': clear_area,
        'trees_small_nos': tree_g1,
        'trees_large_nos': tree_g2,
        'stumps_nos': stumps,
        'build_m3': build_demol,
        'pipe_m': pipe_rem,
        # Class E Earthworks
        'veg_exc_m3': veg_vol,
        'mass_exc_m3': total_excav,
        'disp_veg_m3': disposal_veg,
        'disp_other_m3': disposal_other,
        'prep_exc_m2': prep_area,
        'murram_m3': total_murram,
        'hardcore_m2': hardcore_area,
        'topsoil_m3': topsoil_vol,
        'trim_ts_m2': trim_topsoil,
        'prep_fill_m2': prep_filled_other,
        # Class R
        'bitumen_m2': bitumen_area,
        'kerb_straight_m': kerb_straight,
        'kerb_curved_m': kerb_curved,
        # Landscaping
        'grass_m2': grass_seed_area,
        'mahogany_nos': mahogany_nos,
        'hedge_m': hedge_len_net,
        # Class X
        'fence1_m': fence1_len,
        'fence2_m': fence2_len,
        'gates_nos': gates_nos
    }

if __name__ == "__main__":
    quantities = calculate_roadworks_takeoff()
    # Optional: Export to CSV
    import csv
    with open('roadworks_boq.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=quantities.keys())
        writer.writeheader()
        writer.writerow(quantities)
    print("\nExported to roadworks_boq.csv")
```

```python
# substructure_takeoff_app.py
import math

def calculate_substructure_takeoff():
    print("Substructure Quantity Takeoff Calculator (Flexible for Strip, Columns, Recesses, Voids)")
    
    # Core Inputs
    plan_type = input("Plan type (rectangle/semi-circle/complex): ").lower()
    ext_length = float(input("External length (m): "))
    ext_width = float(input("External width (m): "))
    has_internal_walls = input("Has internal walls? (y/n): ").lower() == 'y'
    if has_internal_walls:
        int_wall_len = float(input("Total internal wall length (m): "))
    else:
        int_wall_len = 0.0
    
    # Variations
    has_columns = input("Has columns? (y/n): ").lower() == 'y'
    if has_columns:
        num_columns = int(input("Number of columns: "))
        col_size = float(input("Column size (m, square side): "))
        col_base_size = float(input("Column base size (m, square): "))
        col_excav_depth = float(input("Column excavation depth (m): "))
    else:
        num_columns = col_size = col_base_size = col_excav_depth = 0
    
    has_recess = input("Has recess/void? (y/n): ").lower() == 'y'
    if has_recess:
        recess_type = input("Recess type (corner/center/bay): ").lower()
        recess_len = float(input("Recess length (m): "))
        recess_wid = float(input("Recess width (m): "))
    else:
        recess_len = recess_wid = 0.0
        recess_type = ''
    
    has_cavity_wall = input("Cavity walls? (y/n): ").lower() == 'y'
    cavity_thick = float(input("Cavity thickness if any (m): ")) if has_cavity_wall else 0.0
    
    # Depths and Thicknesses
    veg_depth = float(input("Vegetable soil depth (m): "))
    trench_depth = float(input("Trench excavation depth (m): "))
    reduce_level_depth = float(input("Reduce level depth (m): "))
    wall_thick = float(input("Foundation wall thickness (m): "))
    conc_thick_strip = float(input("Strip concrete thickness (m): "))
    hardcore_thick = float(input("Hardcore thickness (m): "))
    blinding_thick = float(input("Blinding thickness (m): "))
    dpm_thick = float(input("DPM thickness (mm): ")) / 1000
    anti_termite = input("Anti-termite treatment? (y/n): ").lower() == 'y'
    has_formwork = input("Formwork? (y/n): ").lower() == 'y'
    has_reinforce = input("Reinforcement? (y/n): ").lower() == 'y'
    if has_reinforce:
        rebar_len = float(input("Total rebar length (m): "))
    else:
        rebar_len = 0
    
    # Clearances and Areas
    clear_extra = float(input("Extra clearance each side (m): "))
    site_clear_area = (ext_length + 2*clear_extra) * (ext_width + 2*clear_extra)
    if has_recess:
        if recess_type == 'corner':
            site_clear_area -= recess_len * recess_wid
        elif recess_type == 'center':
            site_clear_area -= 0.5 * recess_len * recess_wid  # Adjust based on video logic
        elif recess_type == 'bay':
            site_clear_area += (3/14) * math.pi * (recess_len/2)**2  # Approx sector
    
    veg_vol = site_clear_area * veg_depth
    
    # Excavation
    center_line_wall = 2*(ext_length + ext_width) + int_wall_len - (4*wall_thick if plan_type == 'rectangle' else 0)
    if has_recess:
        if recess_type == 'corner':
            center_line_wall -= 2 * recess_wid  # Adjust girth
        elif recess_type == 'center':
            center_line_wall += 2 * recess_len  # Add projections
    trench_excav_vol = center_line_wall * wall_thick * trench_depth
    reduce_level_vol = site_clear_area * reduce_level_depth
    
    col_excav_vol = num_columns * (col_base_size**2) * col_excav_depth if has_columns else 0
    total_excav = trench_excav_vol + reduce_level_vol + col_excav_vol
    
    # Disposal
    disposal_veg = veg_vol
    disposal_other = total_excav - veg_vol  # Assume backfill some
    
    # Concrete
    conc_strip_vol = center_line_wall * wall_thick * conc_thick_strip
    conc_col_vol = num_columns * (col_base_size**2) * conc_thick_strip if has_columns else 0
    total_conc = conc_strip_vol + conc_col_vol
    
    # Walling
    wall_height = trench_depth - conc_thick_strip
    wall_area = center_line_wall * wall_height
    if has_cavity_wall:
        wall_area *= 2  # Inner/outer
        wall_area -= center_line_wall * cavity_thick * wall_height  # Cavity void
    
    # Backfill
    backfill_soil_vol = (total_excav - total_conc - wall_area * wall_thick) * 0.5  # Approx half reused
    hardcore_vol = site_clear_area * hardcore_thick
    if has_recess:
        hardcore_vol -= recess_len * recess_wid * hardcore_thick
    
    # Blinding/DPM/Anti-termite
    blinding_area = site_clear_area
    dpm_area = blinding_area
    anti_termite_area = blinding_area if anti_termite else 0
    
    # Formwork
    formwork_area = 0
    if has_formwork:
        formwork_perim = 2*(ext_length + ext_width)
        formwork_area = formwork_perim * conc_thick_strip * 2  # Both sides
    
    # Reinforcement
    reinforce_len = rebar_len if has_reinforce else 0
    
    # Reinstating Topsoil
    reinstate_perim = 2*(ext_length + ext_width + (col_base_size * num_columns if has_columns else 0))
    reinstate_width = 0.4  # Default from videos
    reinstate_depth = veg_depth
    reinstate_vol = (reinstate_perim / 2) * reinstate_width * reinstate_depth  # Center line approx
    
    # Output
    print("\n--- QUANTITIES ---")
    print(f"Site Clearance: {site_clear_area:.2f} m²")
    print(f"Veg Soil Excavation: {veg_vol:.2f} m³")
    print(f"Trench/Reduce Excavation: {total_excav:.2f} m³")
    print(f"Disposal Veg: {disposal_veg:.2f} m³")
    print(f"Disposal Other: {disposal_other:.2f} m³")
    print(f"Foundation Concrete: {total_conc:.2f} m³")
    print(f"Foundation Walling: {wall_area:.2f} m²")
    print(f"Backfill Soil: {backfill_soil_vol:.2f} m³")
    print(f"Hardcore: {hardcore_vol:.2f} m³")
    print(f"Blinding: {blinding_area:.2f} m²")
    print(f"DPM: {dpm_area:.2f} m²")
    if anti_termite:
        print(f"Anti-termite: {anti_termite_area:.2f} m²")
    if has_formwork:
        print(f"Formwork: {formwork_area:.2f} m²")
    if has_reinforce:
        print(f"Reinforcement: {reinforce_len:.2f} m")
    print(f"Reinstating Topsoil: {reinstate_vol:.2f} m³")
    
    return {
        'clear_m2': site_clear_area,
        'veg_exc_m3': veg_vol,
        'excav_m3': total_excav,
        'disp_veg_m3': disposal_veg,
        'disp_other_m3': disposal_other,
        'conc_m3': total_conc,
        'wall_m2': wall_area,
        'backfill_m3': backfill_soil_vol,
        'hardcore_m3': hardcore_vol,
        'blinding_m2': blinding_area,
        'dpm_m2': dpm_area,
        'anti_termite_m2': anti_termite_area,
        'formwork_m2': formwork_area,
        'reinforce_m': reinforce_len,
        'reinstate_m3': reinstate_vol
    }

if __name__ == "__main__":
    quantities = calculate_substructure_takeoff()
```

```python
# drainage_takeoff_app.py
import math

def calculate_drainage_takeoff():
    print("Drainage Works Quantity Takeoff Calculator (Manholes, Pipes, Flexible Variations)")
    
    # General Inputs
    veg_depth = float(input("Vegetable soil depth (m, default 0.15): ") or 0.15)
    has_rock = input("Extra over for rock? (y/n): ").lower() == 'y'
    if has_rock:
        rock_start_depth = float(input("Rock starts at depth from GL (m): "))
    else:
        rock_start_depth = 0
    has_planking = input("Planking/strutting/dewatering? (y/n): ").lower() == 'y'
    
    # Manholes
    has_manholes = input("Include manholes? (y/n): ").lower() == 'y'
    if has_manholes:
        num_mh = int(input("Number of manholes: "))
        mh_data = []
        total_excav_veg_m2 = total_excav_pit_m3 = total_rock_m3 = 0
        total_conc_bed_m3 = total_wall_m2 = total_bench_nos = total_channel_m = 0
        total_slab_m3 = total_form_soffit_m2 = total_form_edge_m = total_plaster_m2 = 0
        total_cover_nos = total_backfill_m3 = 0
        
        for i in range(1, num_mh + 1):
            print(f"\nManhole {i}:")
            mh_type = input("Type (rect/circ): ").lower()
            if mh_type == 'rect':
                int_l = float(input("Internal length (m): "))
                int_w = float(input("Internal width (m): "))
                int_diam = 0
            else:
                int_diam = float(input("Internal diameter (m): "))
                int_l = int_w = int_diam
            invert = float(input("Invert level (m): "))
            bed_thick = float(input("Bed thickness (m): "))
            wall_thick = float(input("Wall thickness (m): "))
            proj_thick = float(input("Projection thickness (m): "))
            slab_thick = float(input("Slab thickness (m): "))
            bench_avg_h = float(input("Benching avg height (m, default 0.2): ") or 0.2)
            has_bench = bench_avg_h > 0
            plaster_thick = float(input("Plaster thickness (mm, 0 if none): ")) / 1000
            has_plaster = plaster_thick > 0
            cover_l = float(input("Cover length (m): "))
            cover_w = float(input("Cover width (m): "))
            has_channel = input("Has channel? (y/n): ").lower() == 'y'
            if has_channel:
                channel_len = float(input("Channel length (m): "))
            else:
                channel_len = 0
            num_this_mh = int(input(f"Number of this manhole type: "))
            
            # Excav Area
            ext_l = int_l + 2 * (wall_thick + proj_thick)
            ext_w = int_w + 2 * (wall_thick + proj_thick)
            if mh_type == 'circ':
                ext_diam = int_diam + 2 * (wall_thick + proj_thick)
                excav_area = math.pi * (ext_diam / 2)**2
            else:
                excav_area = ext_l * ext_w
            
            # Veg Excav
            veg_excav_m2 = excav_area * num_this_mh
            total_excav_veg_m2 += veg_excav_m2
            
            # Pit Depth
            pit_depth = invert + bed_thick + 0.05  # Extra as per vids
            excav_pit_m3 = excav_area * pit_depth * num_this_mh
            
            # Stages if >1.5m
            if pit_depth > 1.5:
                stage1_depth = 1.5
                stage2_depth = pit_depth - 1.5
                print(f"Excav pit stage1 (<=1.5m): {excav_area * stage1_depth * num_this_mh:.2f} m³")
                print(f"Excav pit stage2 (>1.5m): {excav_area * stage2_depth * num_this_mh:.2f} m³")
            total_excav_pit_m3 += excav_pit_m3
            
            # Rock
            if has_rock:
                rock_depth = max(0, pit_depth - rock_start_depth)
                rock_m3 = excav_area * rock_depth * num_this_mh
                total_rock_m3 += rock_m3
            
            # Concrete Bed
            conc_bed_m3 = excav_area * bed_thick * num_this_mh
            total_conc_bed_m3 += conc_bed_m3
            
            # Wall Centerline/Girth
            if mh_type == 'rect':
                int_perim = 2 * (int_l + int_w)
                centerline = int_perim + 2 * wall_thick  # As per vids: int_perim + 4*(thick/2)
            else:
                centerline = math.pi * (int_diam + wall_thick)
            wall_height = invert + 0.05 - slab_thick
            wall_m2 = centerline * wall_height * num_this_mh
            total_wall_m2 += wall_m2
            
            # Benching
            if has_bench:
                if mh_type == 'rect':
                    bench_area = int_l * int_w
                else:
                    bench_area = math.pi * (int_diam / 2)**2
                bench_nos = num_this_mh  # Often nos
                total_bench_nos += bench_nos
            
            # Channel
            total_channel_m += channel_len * num_this_mh
            
            # Slab
            if mh_type == 'rect':
                slab_area = ext_l * ext_w - cover_l * cover_w
            else:
                slab_area = math.pi * (ext_diam / 2)**2 - cover_l * cover_w
            slab_m3 = slab_area * slab_thick * num_this_mh
            total_slab_m3 += slab_m3
            
            # Formwork Soffit
            if mh_type == 'rect':
                soffit_m2 = int_l * int_w * num_this_mh
            else:
                soffit_m2 = math.pi * (int_diam / 2)**2 * num_this_mh
            total_form_soffit_m2 += soffit_m2
            
            # Formwork Edges
            if mh_type == 'rect':
                edge_perim = 2 * (ext_l + ext_w)
            else:
                edge_perim = math.pi * ext_diam
            edge_m = edge_perim * num_this_mh
            deduct_edge = 2 * (cover_l + cover_w) * num_this_mh  # Approx for rect covers
            total_form_edge_m += edge_m - deduct_edge
            
            # Plaster
            if has_plaster:
                int_girth = int_perim if mh_type == 'rect' else math.pi * int_diam
                plaster_wall_m2 = int_girth * wall_height * num_this_mh
                plaster_bench_m2 = bench_area * num_this_mh if has_bench else 0
                total_plaster_m2 += plaster_wall_m2 + plaster_bench_m2
            
            # Cover
            total_cover_nos += num_this_mh
            
            # Backfill
            conc_vol = conc_bed_m3 + slab_m3 + (wall_m2 * wall_thick)  # Approx
            backfill_m3 = excav_pit_m3 - conc_vol
            total_backfill_m3 += backfill_m3
        
        # Output Manholes
        print("\n--- MANHOLE QUANTITIES ---")
        print(f"Veg Excav: {total_excav_veg_m2:.2f} m²")
        print(f"Pit Excav: {total_excav_pit_m3:.2f} m³")
        if has_rock: print(f"Rock Extra: {total_rock_m3:.2f} m³")
        print(f"Conc Bed: {total_conc_bed_m3:.2f} m³")
        print(f"Walling: {total_wall_m2:.2f} m²")
        if has_bench: print(f"Benching: {total_bench_nos} nos")
        print(f"Channel: {total_channel_m:.2f} m")
        print(f"Slab Conc: {total_slab_m3:.2f} m³")
        print(f"Form Soffit: {total_form_soffit_m2:.2f} m²")
        print(f"Form Edges: {total_form_edge_m:.2f} m")
        if has_plaster: print(f"Plaster: {total_plaster_m2:.2f} m²")
        print(f"Covers: {total_cover_nos} nos")
        print(f"Backfill: {total_backfill_m3:.2f} m³")
        if has_planking: print("Planking/Strutting/Dewatering: Item")
    
    # Pipes
    has_pipes = input("Include pipes? (y/n): ").lower() == 'y'
    if has_pipes:
        num_runs = int(input("Number of pipe runs: "))
        total_trench_m3 = total_bed_m3 = total_pipe_m = total_backfill_pipe_m3 = 0
        
        for i in range(1, num_runs + 1):
            print(f"\nPipe Run {i}:")
            pipe_len = float(input("Pipe length (m): "))
            diam_mm = float(input("Diameter (mm): "))
            diam = diam_mm / 1000
            trench_depth = float(input("Trench avg depth (m): "))
            trench_wid = float(input("Trench width (m): "))
            surround_thick = float(input("Bedding/surround thick (m): "))
            num_this = int(input("Number of this run: "))
            
            trench_m3 = pipe_len * trench_wid * trench_depth * num_this
            total_trench_m3 += trench_m3
            
            bed_area = math.pi * ((diam/2 + surround_thick)**2 - (diam/2)**2)
            bed_m3 = pipe_len * bed_area * num_this
            total_bed_m3 += bed_m3
            
            pipe_m = pipe_len * num_this
            total_pipe_m += pipe_m
            
            pipe_vol = pipe_len * math.pi * (diam/2)**2 * num_this
            backfill_pipe_m3 = trench_m3 - bed_m3 - pipe_vol
            total_backfill_pipe_m3 += backfill_pipe_m3
        
        # Output Pipes
        print("\n--- PIPE QUANTITIES ---")
        print(f"Trench Excav: {total_trench_m3:.2f} m³")
        print(f"Bedding: {total_bed_m3:.2f} m³")
        print(f"Pipes: {total_pipe_m:.2f} m")
        print(f"Backfill: {total_backfill_pipe_m3:.2f} m³")
        print("Testing/Connections: Item")
    
    # Return Dict
    return {
        'veg_excav_m2': total_excav_veg_m2 if has_manholes else 0,
        'pit_excav_m3': total_excav_pit_m3 if has_manholes else 0,
        'rock_m3': total_rock_m3 if has_rock and has_manholes else 0,
        'conc_bed_m3': total_conc_bed_m3 if has_manholes else 0,
        'wall_m2': total_wall_m2 if has_manholes else 0,
        'bench_nos': total_bench_nos if has_bench and has_manholes else 0,
        'channel_m': total_channel_m if has_manholes else 0,
        'slab_m3': total_slab_m3 if has_manholes else 0,
        'form_soffit_m2': total_form_soffit_m2 if has_manholes else 0,
        'form_edge_m': total_form_edge_m if has_manholes else 0,
        'plaster_m2': total_plaster_m2 if has_plaster and has_manholes else 0,
        'covers_nos': total_cover_nos if has_manholes else 0,
        'backfill_m3': total_backfill_m3 if has_manholes else 0,
        'trench_m3': total_trench_m3 if has_pipes else 0,
        'bedding_m3': total_bed_m3 if has_pipes else 0,
        'pipe_m': total_pipe_m if has_pipes else 0,
        'backfill_pipe_m3': total_backfill_pipe_m3 if has_pipes else 0
    }

if __name__ == "__main__":
    quantities = calculate_drainage_takeoff()
```

```python
# roof_takeoff_app.py
import math

def calculate_roof_takeoff():
    print("Roof Quantity Takeoff Calculator (Gable/Hipped/Pitched, Civil Precision)")
    
    # Core Inputs
    roof_type = input("Roof type (gable/hipped/pitched): ").lower()
    build_len = float(input("Building length (m): "))
    build_wid = float(input("Building width/span (m): "))
    wall_thick = float(input("Wall thickness (m): "))
    overhang = float(input("Eaves overhang (m): "))
    pitch_deg = float(input("Pitch angle (degrees): "))
    truss_space = float(input("Truss spacing (m): "))
    rafter_space = float(input("Common rafter spacing (m): "))
    bearer_space = float(input("Bearer spacing (m): "))
    
    # Thicknesses/Sizes
    wp_thick = float(input("Wall plate size (e.g., 0.1x0.05, enter length dim): "))
    rafter_thick = float(input("Rafter thickness (m): "))
    tie_thick = float(input("Tie beam thickness (m): "))
    strut_thick = float(input("Strut/tie thickness (m): "))
    binder_thick = float(input("Binder thickness (m): "))
    ridge_thick = float(input("Ridge board thickness (m): "))
    fascia_thick = float(input("Fascia thickness (m): "))
    gutter_wid = float(input("Gutter width (m): "))
    tile_lap = float(input("Tile lap (m): "))
    downpipe_len = float(input("Downpipe length (m): "))
    
    # Variations
    has_hip = roof_type == 'hipped'
    if has_hip:
        hip_len = math.sqrt((build_len/2)**2 + (build_wid/2)**2) / math.cos(math.radians(pitch_deg))
        num_hips = 4
    else:
        hip_len = num_hips = 0
    has_valley = input("Has valleys? (y/n): ").lower() == 'y'
    if has_valley:
        valley_len = float(input("Valley length (m): "))
    else:
        valley_len = 0
    
    has_covering = input("Include covering? (y/n): ").lower() == 'y'
    has_eaves_verges = input("Include eaves/verges? (y/n): ").lower() == 'y'
    has_rainwater = input("Include rainwater goods? (y/n): ").lower() == 'y'
    
    # Calculations
    # Wall Plate
    wp_len = build_len + 2 * wall_thick
    num_wp = 2  # Long sides
    wp_m = wp_len * num_wp
    
    # Number of Trusses
    num_truss = math.floor(build_len / truss_space) + 1
    
    # Rafter Length
    half_span = (build_wid + 2 * wall_thick + 2 * overhang - 2 * fascia_thick - gutter_wid) / 2
    rafter_len = half_span / math.cos(math.radians(pitch_deg))
    princ_rafter_m = rafter_len * 2 * num_truss  # 2 per truss
    
    # Common Rafters
    num_common = math.floor(build_len / rafter_space) + 1 - num_truss  # Exclude principals
    common_rafter_m = rafter_len * 2 * num_common  # 2 sides
    
    # Tie Beam
    tie_len = build_wid + 2 * wall_thick + rafter_thick
    tie_m = tie_len * num_truss
    
    # Ties and Struts
    c = half_span * math.cos(math.radians(pitch_deg))
    half_c = c / 2
    tie_a = half_c * math.tan(math.radians(pitch_deg))
    strut_b = half_c / math.cos(math.radians(pitch_deg))
    ties_m = tie_a * 2 * num_truss  # 2 per truss
    struts_m = strut_b * 2 * num_truss
    
    # Binder/Ridge
    binder_len = build_len + 2 * wall_thick + 2 * overhang - 2 * fascia_thick - gutter_wid
    binder_m = binder_len * 2  # 2 sides
    ridge_m = binder_len
    
    # Hips/Valleys
    hips_m = hip_len * num_hips if has_hip else 0
    valleys_m = valley_len if has_valley else 0
    
    # Covering
    if has_covering:
        cover_len = build_len + 2 * wall_thick + 2 * overhang
        cover_wid = rafter_len * 2
        if has_hip:
            cover_area = (build_len * build_wid) + adjustments  # Approx trapezoid, but use video logic
        else:
            cover_area = cover_len * cover_wid
        tiles_m2 = cover_area / (1 - tile_lap / tile_size) if 'tile_size' defined else cover_area
        ridge_cover_m = cover_len
        ridge_ends_nos = 2
    
    # Eaves/Verges
    if has_eaves_verges:
        fascia_len = cover_len - 0.25  # Adjust per video
        fascia_m = fascia_len * 2
        verge_len = rafter_len
        verge_m = verge_len * 4  # 2 per gable end
        runner_len = build_len + 2 * wall_thick
        runner_m = runner_len * 2
        num_bearers = math.floor(build_len / bearer_space) + 1
        bearer_len = overhang - fascia_thick - gutter_wid
        bearers_m = bearer_len * num_bearers * 2
        soffit_area = (cover_len - 0.25) * bearer_len  # T&G
        verge_soffit_area = verge_len * bearer_len * 2
    
    # Rainwater
    if has_rainwater:
        gutter_len = cover_len * 2  # 2 sides
        gutter_ends_nos = 4
        outlets_nos = 4
        downpipe_m = downpipe_len * 4  # Assume 4
        shoes_nos = 4
    
    # Output
    print("\n--- ROOF QUANTITIES ---")
    print(f"Wall Plate: {wp_m:.2f} m (100x50mm)")
    print(f"Principal Rafters: {princ_rafter_m:.2f} m (150x50mm)")
    print(f"Common Rafters: {common_rafter_m:.2f} m (150x50mm)")
    print(f"Tie Beams: {tie_m:.2f} m (150x50mm)")
    print(f"Ties: {ties_m:.2f} m (100x50mm)")
    print(f"Struts: {struts_m:.2f} m (100x50mm)")
    print(f"Binders: {binder_m:.2f} m (150x50mm)")
    print(f"Ridge Board: {ridge_m:.2f} m (175x25mm)")
    if has_hip: print(f"Hips: {hips_m:.2f} m")
    if has_valley: print(f"Valleys: {valleys_m:.2f} m")
    
    if has_covering:
        print(f"Covering Area: {cover_area:.2f} m² (tiles with lap)")
        print(f"Ridge Cover: {ridge_cover_m:.2f} m, Ends: {ridge_ends_nos} nos")
    
    if has_eaves_verges:
        print(f"Fascia: {fascia_m:.2f} m (200x25mm)")
        print(f"Verges: {verge_m:.2f} m")
        print(f"Runners: {runner_m:.2f} m (50x50mm)")
        print(f"Bearers: {bearers_m:.2f} m (50x50mm)")
        print(f"Soffit T&G: {soffit_area:.2f} m² (25mm)")
        print(f"Verge Soffit: {verge_soffit_area:.2f} m²")
    
    if has_rainwater:
        print(f"Gutters: {gutter_len:.2f} m (100mm)")
        print(f"Gutter Ends: {gutter_ends_nos} nos, Outlets: {outlets_nos} nos")
        print(f"Downpipes: {downpipe_m:.2f} m (75mm), Shoes: {shoes_nos} nos")
    
    return {
        'wp_m': wp_m,
        'princ_rafter_m': princ_rafter_m,
        'common_rafter_m': common_rafter_m,
        'tie_m': tie_m,
        'ties_m': ties_m,
        'struts_m': struts_m,
        'binder_m': binder_m,
        'ridge_m': ridge_m,
        'hips_m': hips_m,
        'valleys_m': valleys_m,
        'cover_m2': cover_area if has_covering else 0,
        'ridge_cover_m': ridge_cover_m if has_covering else 0,
        'fascia_m': fascia_m if has_eaves_verges else 0,
        'verge_m': verge_m if has_eaves_verges else 0,
        'runner_m': runner_m if has_eaves_verges else 0,
        'bearers_m': bearers_m if has_eaves_verges else 0,
        'soffit_m2': soffit_area if has_eaves_verges else 0,
        'gutter_m': gutter_len if has_rainwater else 0,
        'downpipe_m': downpipe_m if has_rainwater else 0
    }

if __name__ == "__main__":
    quantities = calculate_roof_takeoff()
```

```python
# superstructure_takeoff_app.py
import math

def calculate_superstructure_takeoff():
    print("Superstructure Quantity Takeoff Calculator (Walls, Beams, Slabs, Columns, Parapet, Rainwater)")
    
    # Core Inputs
    plan_type = input("Plan type (rect/complex): ").lower()
    ext_len = float(input("External length (m): "))
    ext_wid = float(input("External width (m): "))
    wall_height = float(input("Wall height (m): "))
    wall_thick = float(input("Wall thickness (m): "))
    has_internal_walls = input("Internal walls? (y/n): ").lower() == 'y'
    if has_internal_walls:
        int_wall_total_len = float(input("Total internal wall length (m): "))
    else:
        int_wall_total_len = 0
    
    # Openings
    has_openings = input("Has openings (doors/windows)? (y/n): ").lower() == 'y'
    if has_openings:
        num_doors = int(input("Number of doors: "))
        door_area = float(input("Avg door area (m²): "))
        num_windows = int(input("Number of windows: "))
        win_area = float(input("Avg window area (m²): "))
    else:
        num_doors = door_area = num_windows = win_area = 0
    
    # Columns
    has_columns = input("Has columns? (y/n): ").lower() == 'y'
    if has_columns:
        num_cols = int(input("Number of columns: "))
        col_size = float(input("Column size (m, square side): "))
        col_height = float(input("Column height (m): "))
        has_form_col = input("Formwork for columns? (y/n): ").lower() == 'y'
    else:
        num_cols = col_size = col_height = 0
        has_form_col = False
    
    # Beams
    has_beams = input("Has beams? (y/n): ").lower() == 'y'
    if has_beams:
        beam_len_total = float(input("Total beam length (m): "))
        beam_wid = float(input("Beam width (m): "))
        beam_depth = float(input("Beam depth (m): "))
        has_form_beam = input("Formwork for beams? (y/n): ").lower() == 'y'
    else:
        beam_len_total = beam_wid = beam_depth = 0
        has_form_beam = False
    
    # Slab
    has_slab = input("Has slab? (y/n): ").lower() == 'y'
    if has_slab:
        slab_area = float(input("Slab area (m²): "))
        slab_thick = float(input("Slab thickness (m): "))
        has_form_slab = input("Formwork for slab soffit? (y/n): ").lower() == 'y'
        has_reinf_slab = input("Reinforcement in slab? (y/n): ").lower() == 'y'
        if has_reinf_slab:
            rebar_slab_kg = float(input("Rebar in slab (kg/m³): "))
        else:
            rebar_slab_kg = 0
    else:
        slab_area = slab_thick = 0
        has_form_slab = has_reinf_slab = False
        rebar_slab_kg = 0
    
    # Parapet
    has_parapet = input("Has parapet wall? (y/n): ").lower() == 'y'
    if has_parapet:
        parapet_girth = float(input("Parapet girth (m): "))
        parapet_height = float(input("Parapet height (m): "))
        parapet_thick = float(input("Parapet thickness (m): "))
        has_coping = input("Has coping? (y/n): ").lower() == 'y'
        if has_coping:
            coping_wid = float(input("Coping width (m): "))
            coping_thick = float(input("Coping thickness (m): "))
        else:
            coping_wid = coping_thick = 0
    else:
        parapet_girth = parapet_height = parapet_thick = 0
        has_coping = False
        coping_wid = coping_thick = 0
    
    # Rainwater Goods
    has_rainwater = input("Has rainwater goods? (y/n): ").lower() == 'y'
    if has_rainwater:
        downpipe_len = float(input("Downpipe length (m): "))
        num_downpipes = int(input("Number of downpipes: "))
        outlet_diam = float(input("Outlet diameter (mm): ")) / 1000
        has_shoe = input("Has shoe? (y/n): ").lower() == 'y'
        if has_shoe:
            shoe_len = float(input("Shoe length (m): "))
        else:
            shoe_len = 0
    else:
        downpipe_len = num_downpipes = outlet_diam = shoe_len = 0
        has_shoe = False
    
    # Wall Calculations
    ext_perim = 2 * (ext_len + ext_wid)
    centerline_wall = ext_perim - 4 * wall_thick + int_wall_total_len  # Adjust corners
    wall_area_gross = centerline_wall * wall_height
    openings_area = (num_doors * door_area) + (num_windows * win_area)
    wall_area_net = wall_area_gross - openings_area
    
    # Columns
    col_conc_vol = num_cols * (col_size ** 2) * col_height
    form_col_m2 = num_cols * 4 * col_size * col_height if has_form_col else 0
    
    # Beams
    beam_conc_vol = beam_len_total * beam_wid * beam_depth
    form_beam_m2 = beam_len_total * (2 * beam_depth + beam_wid) if has_form_beam else 0  # Sides + bottom
    
    # Slab
    slab_conc_vol = slab_area * slab_thick
    form_slab_m2 = slab_area if has_form_slab else 0
    rebar_slab_total = slab_conc_vol * rebar_slab_kg
    
    # Parapet
    parapet_area = parapet_girth * parapet_height
    coping_vol = parapet_girth * coping_wid * coping_thick if has_coping else 0
    
    # Rainwater
    downpipe_m = downpipe_len * num_downpipes
    outlets_nos = num_downpipes
    shoes_nos = num_downpipes if has_shoe else 0
    
    # Output
    print("\n--- QUANTITIES ---")
    print(f"Wall Area Net: {wall_area_net:.2f} m² (blockwork)")
    if has_columns:
        print(f"Column Concrete: {col_conc_vol:.2f} m³")
        if has_form_col: print(f"Column Formwork: {form_col_m2:.2f} m²")
    if has_beams:
        print(f"Beam Concrete: {beam_conc_vol:.2f} m³")
        if has_form_beam: print(f"Beam Formwork: {form_beam_m2:.2f} m²")
    if has_slab:
        print(f"Slab Concrete: {slab_conc_vol:.2f} m³")
        if has_form_slab: print(f"Slab Formwork: {form_slab_m2:.2f} m²")
        if has_reinf_slab: print(f"Slab Rebar: {rebar_slab_total:.2f} kg")
    if has_parapet:
        print(f"Parapet Wall: {parapet_area:.2f} m²")
        if has_coping: print(f"Coping Concrete: {coping_vol:.2f} m³")
    if has_rainwater:
        print(f"Downpipes: {downpipe_m:.2f} m ({outlet_diam*1000:.0f}mm)")
        print(f"Outlets: {outlets_nos} nos")
        if has_shoe: print(f"Shoes: {shoes_nos} nos ({shoe_len:.2f} m each)")
    
    return {
        'wall_m2': wall_area_net,
        'col_conc_m3': col_conc_vol,
        'col_form_m2': form_col_m2,
        'beam_conc_m3': beam_conc_vol,
        'beam_form_m2': form_beam_m2,
        'slab_conc_m3': slab_conc_vol,
        'slab_form_m2': form_slab_m2,
        'slab_rebar_kg': rebar_slab_total,
        'parapet_m2': parapet_area,
        'coping_m3': coping_vol,
        'downpipe_m': downpipe_m,
        'outlets_nos': outlets_nos,
        'shoes_nos': shoes_nos
    }

if __name__ == "__main__":
    quantities = calculate_superstructure_takeoff()
```

```python
# underground_tank_takeoff.py
import math

def calculate_underground_tank():
    print("Underground Tank Quantity Takeoff Calculator (Circular/Square/Rectangular)")
    
    # Tank type
    tank_type = input("Tank type (circular/square/rectangular): ").lower()
    
    # Common inputs
    veg_soil = float(input("Vegetable soil removal depth (m, default 0.15): ") or 0.15)
    working_space = float(input("Working space each side (m, default 0.3): ") or 0.3)
    blinding_thick = float(input("Blinding thickness (m, default 0.075): ") or 0.075)
    bed_thick = float(input("Concrete bed thickness (m, default 0.15): ") or 0.15)
    wall_thick = float(input("Wall thickness (m, default 0.2): ") or 0.2)
    slab_thick = float(input("Top slab thickness (m, default 0.15): ") or 0.15)
    plaster_thick = float(input("Internal plaster thickness (m, default 0.012): ") or 0.012)
    mastic_thick = float(input("Mastic asphalt thickness (m, default 0.02): ") or 0.02)
    brick_thick = float(input("Protective brick layer thickness (m, default 0.115): ") or 0.115)
    cover_to_slab = float(input("Cover depth above slab (m, default 0.3): ") or 0.3)
    num_covers = int(input("Number of manhole covers: ") or 1)
    cover_L = float(input("Cover length (m, default 0.6): ") or 0.6)
    cover_W = float(input("Cover width (m, default 0.6): ") or 0.6)
    
    if tank_type == "circular":
        int_diam = float(input("Internal diameter (m): "))
        int_radius = int_diam / 2
        depth_int = float(input("Internal depth (m): "))
        
        # External dimensions
        ext_diam = int_diam + 2*(plaster_thick + wall_thick + mastic_thick + brick_thick)
        ext_radius = ext_diam / 2
        
        # Areas
        int_area = math.pi * int_radius ** 2
        ext_area = math.pi * ext_radius ** 2
        
        # Excavation dims
        excav_diam = ext_diam + 2*working_space
        excav_radius = excav_diam / 2
        excav_area = math.pi * excav_radius ** 2
        excav_depth = depth_int + bed_thick + blinding_thick + cover_to_slab + veg_soil
        net_excav_depth = excav_depth - veg_soil
        
        # Volumes
        excavation = excav_area * net_excav_depth
        topsoil = excav_area * veg_soil
        blinding = excav_area
        concrete_bed = ext_area * bed_thick
        wall_height = depth_int
        wall_girth = math.pi * int_diam
        wall_vol = wall_girth * wall_height * wall_thick
        slab_area_gross = ext_area
        openings = num_covers * cover_L * cover_W
        slab_net = slab_area_gross - openings
        slab_conc = slab_net * slab_thick
        backfill_vol = excavation - (concrete_bed + wall_vol + slab_conc + blinding_thick * excav_area)
        formwork_soffit = int_area
        formwork_edges = num_covers * 2 * (cover_L + cover_W) * slab_thick
        formwork = formwork_soffit + formwork_edges + (math.pi * int_diam * slab_thick)
        
    elif tank_type in ["square", "rectangular"]:
        int_L = float(input("Internal length (m): "))
        int_W = float(input("Internal width (m): "))
        if tank_type == "square":
            int_W = int_L
        depth_int = float(input("Internal depth (m): "))
        
        # External dimensions
        ext_L = int_L + 2*(plaster_thick + wall_thick + mastic_thick + brick_thick)
        ext_W = int_W + 2*(plaster_thick + wall_thick + mastic_thick + brick_thick)
        
        # Excavation dims
        excav_L = ext_L + 2*working_space
        excav_W = ext_W + 2*working_space
        excav_area = excav_L * excav_W
        int_area = int_L * int_W
        ext_area = ext_L * ext_W
        excav_depth = depth_int + bed_thick + blinding_thick + cover_to_slab + veg_soil
        net_excav_depth = excav_depth - veg_soil
        
        # Volumes
        excavation = excav_area * net_excav_depth
        topsoil = excav_area * veg_soil
        blinding = excav_area
        concrete_bed = ext_area * bed_thick
        wall_perim = 2 * (int_L + int_W)
        wall_height = depth_int
        wall_vol = wall_perim * wall_height * wall_thick
        slab_area_gross = ext_area
        openings = num_covers * cover_L * cover_W
        slab_net = slab_area_gross - openings
        slab_conc = slab_net * slab_thick
        backfill_vol = excavation - (concrete_bed + wall_vol + slab_conc + blinding_thick * excav_area)
        formwork_soffit = int_area
        formwork_edges = num_covers * 2 * (cover_L + cover_W) * slab_thick
        formwork_perim = 2 * (int_L + int_W) * slab_thick
        formwork = formwork_soffit + formwork_edges + formwork_perim
        
    else:
        print("Invalid tank type.")
        return {}
    
    # Common outputs
    print("\n--- QUANTITIES ---")
    print(f"Site Clearance/Topsoil: {excav_area:.2f} m²")
    print(f"Vegetable Soil Removal: {topsoil:.2f} m³")
    print(f"Excavation (net): {excavation:.2f} m³")
    print(f"Blinding: {blinding:.2f} m²")
    print(f"Concrete Bed: {concrete_bed:.2f} m³")
    print(f"Wall Concrete: {wall_vol:.2f} m³")
    print(f"Slab Concrete: {slab_conc:.2f} m³")
    print(f"Formwork: {formwork:.2f} m²")
    print(f"Backfill: {backfill_vol:.2f} m³")
    
    return {
        'topsoil_m3': topsoil,
        'excavation_m3': excavation,
        'blinding_m2': blinding,
        'bed_m3': concrete_bed,
        'walls_m3': wall_vol,
        'slab_m3': slab_conc,
        'formwork_m2': formwork,
        'backfill_m3': backfill_vol
    }

if __name__ == "__main__":
    quantities = calculate_underground_tank()
```

```python
# basement_takeoff.py
import math

def calculate_basement_takeoff():
    print("Basement Quantity Takeoff Calculator")
    
    # Inputs (all in meters unless specified)
    ext_L = float(input("External length (m, default 3.91): ") or 3.91)
    ext_W = float(input("External width (m, default 3.41): ") or 3.41)
    int_L = float(input("Internal length (m, default 3.0): ") or 3.0)
    int_W = float(input("Internal width (m, default 2.5): ") or 2.5)
    depth_below_gl = float(input("Depth below ground level (m, default 2.5): ") or 2.5)  # From transcripts
    veg_soil_depth = float(input("Vegetable soil removal depth (m, default 0.15): ") or 0.15)
    working_space = float(input("Working space each side (m, default 0.3): ") or 0.3)
    blinding_thick = float(input("Blinding thickness (m, default 0.075): ") or 0.075)
    bed_thick = float(input("Concrete bed thickness (m, default 0.1): ") or 0.1)
    found_L = float(input("Foundation length per side (m, default 0.9): ") or 0.9)
    found_thick = float(input("Foundation thickness (m, default 0.3): ") or 0.3)
    wall_thick = float(input("Wall thickness (m, default 0.225): ") or 0.225)  # Brick wall
    rc_wall_thick = float(input("RC wall thickness (m, default 0.15): ") or 0.15)
    horiz_tanking_thick = float(input("Horizontal tanking thickness (m, default 0.03): ") or 0.03)
    vert_tanking_thick = float(input("Vertical tanking thickness (m, default 0.025): ") or 0.025)
    projection = float(input("Projection (m, default 0.1): ") or 0.1)
    slab_thick = float(input("Ground floor slab thickness (m, default 0.15): ") or 0.15)
    excav_staged = input("Excavate in stages? (yes/no, default no): ").lower() or "no"
    if excav_staged == "yes":
        stage_depth = float(input("Excavation stage depth (m, default 1.5): ") or 1.5)
    else:
        stage_depth = None
    reinf_incl = input("Include reinforcement? (yes/no, default yes): ").lower() or "yes"
    if reinf_incl == "yes":
        reinf_density = float(input("Reinforcement density (kg/m³, default 120): ") or 120)
    form_incl = input("Include formwork? (yes/no, default yes): ").lower() or "yes"
    backfill_incl = input("Include backfill? (yes/no, default yes): ").lower() or "yes"
    disposal_incl = input("Include disposal of excavated material? (yes/no, default yes): ").lower() or "yes"
    planking_item = "Item: Allow for necessary planking and strutting."
    
    # Excavation dimensions
    excav_L = ext_L + 2 * working_space
    excav_W = ext_W + 2 * working_space
    excav_area = excav_L * excav_W
    net_excav_depth = depth_below_gl + bed_thick + blinding_thick + horiz_tanking_thick + found_thick - veg_soil_depth
    excavation_vol = excav_area * net_excav_depth
    
    if excav_staged == "yes":
        num_stages = math.ceil(net_excav_depth / stage_depth)
        staged_vol = []
        remaining_d = net_excav_depth
        for i in range(num_stages):
            d = min(stage_depth, remaining_d)
            vol = excav_area * d
            staged_vol.append(vol)
            remaining_d -= d
    else:
        staged_vol = [excavation_vol]
    
    # Site Clearance (assume bushy as optional in transcripts)
    clearance_area = excav_area
    
    # Vegetable Soil Removal
    veg_soil_vol = clearance_area * veg_soil_depth
    
    # Disposal
    if disposal_incl == "yes":
        disposal_vol = excavation_vol + veg_soil_vol  # All material
    else:
        disposal_vol = 0
    
    # Blinding
    blinding_area = ext_L * ext_W
    
    # Foundation Concrete (footing)
    found_perim = 2 * (ext_L + ext_W)
    found_vol = found_perim * found_L * found_thick / 2  # Approximate for strip footing
    
    # Concrete Bed
    bed_area = ext_L * ext_W
    bed_vol = bed_area * bed_thick
    
    # Horizontal Tanking (asphalt)
    horiz_tanking_area = bed_area
    
    # RC Walls
    rc_wall_perim = 2 * (int_L + int_W)
    rc_wall_height = depth_below_gl
    rc_walls_vol = rc_wall_perim * rc_wall_height * rc_wall_thick
    
    # Vertical Tanking
    vert_tanking_area = rc_wall_perim * rc_wall_height  # One side
    
    # Brick Walls (external and internal)
    ext_wall_perim = 2 * (ext_L + ext_W) - 4 * wall_thick  # Adjust corners
    ext_wall_height = depth_below_gl
    ext_wall_vol = ext_wall_perim * ext_wall_height * wall_thick  # But since brick, perhaps area
    ext_wall_area = ext_wall_perim * ext_wall_height
    
    int_wall_perim = 2 * (int_L + int_W)
    int_wall_height = depth_below_gl
    int_wall_area = int_wall_perim * int_wall_height
    
    # Ground Floor Slab
    slab_area = ext_L * ext_W
    slab_vol = slab_area * slab_thick
    
    # Reinforcement
    if reinf_incl == "yes":
        total_conc_vol = found_vol + bed_vol + rc_walls_vol + slab_vol
        reinf_kg = total_conc_vol * reinf_density
    else:
        reinf_kg = 0
    
    # Formwork
    if form_incl == "yes":
        form_soffit = int_L * int_W  # Base
        form_walls = rc_wall_perim * rc_wall_height * 2  # Inner/outer
        form_slab_edges = 2 * (ext_L + ext_W) * slab_thick
        form_found = found_perim * found_thick * 2  # Sides
        total_form_m2 = form_soffit + form_walls + form_slab_edges + form_found
    else:
        total_form_m2 = 0
    
    # Backfill
    if backfill_incl == "yes":
        backfill_vol = excavation_vol - (bed_vol + rc_walls_vol + slab_vol + found_vol + blinding_thick * blinding_area + horiz_tanking_thick * horiz_tanking_area)
    else:
        backfill_vol = 0
    
    # Outputs
    print("\n--- QUANTITIES ---")
    print(f"Site Clearance: {clearance_area:.2f} m²")
    print(f"Vegetable Soil Removal: {veg_soil_vol:.2f} m³")
    print(f"Basement Excavation: {excavation_vol:.2f} m³")
    if excav_staged == "yes":
        for i, vol in enumerate(staged_vol, 1):
            print(f"Excavation Stage {i}: {vol:.2f} m³")
    print(planking_item)
    if disposal_incl == "yes":
        print(f"Disposal: {disposal_vol:.2f} m³")
    print(f"Blinding: {blinding_area:.2f} m²")
    print(f"Foundation Concrete: {found_vol:.2f} m³")
    print(f"Concrete Bed: {bed_vol:.2f} m³")
    print(f"Horizontal Tanking: {horiz_tanking_area:.2f} m²")
    print(f"RC Walls Concrete: {rc_walls_vol:.2f} m³")
    print(f"Vertical Tanking: {vert_tanking_area:.2f} m²")
    print(f"External Brick Wall: {ext_wall_area:.2f} m²")
    print(f"Internal Brick Wall: {int_wall_area:.2f} m²")
    print(f"Ground Floor Slab Concrete: {slab_vol:.2f} m³")
    if reinf_incl == "yes":
        print(f"Reinforcement: {reinf_kg:.2f} kg")
    if form_incl == "yes":
        print(f"Formwork: {total_form_m2:.2f} m²")
    if backfill_incl == "yes":
        print(f"Backfill: {backfill_vol:.2f} m³")
    
    return {
        'clearance_m2': clearance_area,
        'veg_soil_m3': veg_soil_vol,
        'excavation_m3': excavation_vol,
        'disposal_m3': disposal_vol,
        'blinding_m2': blinding_area,
        'found_conc_m3': found_vol,
        'bed_m3': bed_vol,
        'horiz_tanking_m2': horiz_tanking_area,
        'rc_walls_m3': rc_walls_vol,
        'vert_tanking_m2': vert_tanking_area,
        'ext_wall_m2': ext_wall_area,
        'int_wall_m2': int_wall_area,
        'slab_m3': slab_vol,
        'reinf_kg': reinf_kg,
        'formwork_m2': total_form_m2,
        'backfill_m3': backfill_vol
    }

if __name__ == "__main__":
    quantities = calculate_basement_takeoff()
```

```python
# rcc_superstructure_takeoff.py
import math

def calculate_rcc_superstructure():
    print("RCC Superstructure Takeoff Calculator (Columns, Beams, Slabs)")
    
    # Columns
    num_cols = int(input("Number of columns: "))
    col_dims = []
    col_heights = []
    for i in range(num_cols):
        print(f"\nColumn {i+1}:")
        w = float(input("Width (m): "))
        d = float(input("Depth (m): "))
        h = float(input("Height (m): "))
        col_dims.append((w, d))
        col_heights.append(h)
    
    # Beams
    num_beams = int(input("\nNumber of beams: "))
    beam_lengths = []
    beam_dims = []
    for i in range(num_beams):
        print(f"\nBeam {i+1}:")
        l = float(input("Length (m): "))
        w = float(input("Width (m): "))
        d = float(input("Depth (m): "))
        beam_lengths.append(l)
        beam_dims.append((w, d))
    
    # Slabs
    num_slabs = int(input("\nNumber of slabs: "))
    slab_areas = []
    slab_thick = []
    for i in range(num_slabs):
        print(f"\nSlab {i+1}:")
        area = float(input("Area (m²): "))
        t = float(input("Thickness (m): "))
        slab_areas.append(area)
        slab_thick.append(t)
    
    # Common
    conc_grade = input("Concrete grade (default 1:1.5:3): ") or "1:1.5:3"
    reinf_density = float(input("Reinforcement density (kg/m³, default 120): ") or 120)
    form_type = input("Formwork type (F1-F5, default F3): ") or "F3"
    
    # Calculations
    col_conc_vol = sum(w * d * h for (w, d), h in zip(col_dims, col_heights))
    beam_conc_vol = sum(l * w * d for l, (w, d) in zip(beam_lengths, beam_dims))
    slab_conc_vol = sum(a * t for a, t in zip(slab_areas, slab_thick))
    total_conc_vol = col_conc_vol + beam_conc_vol + slab_conc_vol
    
    # Reinforcement
    col_reinf = col_conc_vol * reinf_density
    beam_reinf = beam_conc_vol * reinf_density
    slab_reinf = slab_conc_vol * reinf_density
    total_reinf = col_reinf + beam_reinf + slab_reinf
    
    # Formwork
    col_form = sum(2 * (w + d) * h for (w, d), h in zip(col_dims, col_heights))
    beam_form_sides = sum(2 * l * d for l, (w, d) in zip(beam_lengths, beam_dims))
    beam_form_bottom = sum(l * w for l, (w, d) in zip(beam_lengths, beam_dims))
    slab_form_soffit = sum(a for a in slab_areas)
    slab_form_edges = sum(2 * (math.sqrt(a) * 4) * t for a, t in zip(slab_areas, slab_thick))  # Approx perimeter
    total_form = col_form + beam_form_sides + beam_form_bottom + slab_form_soffit + slab_form_edges
    
    # Outputs
    print("\n--- QUANTITIES ---")
    print(f"Columns Concrete ({conc_grade}): {col_conc_vol:.3f} m³")
    print(f"Beams Concrete ({conc_grade}): {beam_conc_vol:.3f} m³")
    print(f"Slabs Concrete ({conc_grade}): {slab_conc_vol:.3f} m³")
    print(f"Total Concrete: {total_conc_vol:.3f} m³")
    print(f"Columns Reinforcement: {col_reinf:.1f} kg")
    print(f"Beams Reinforcement: {beam_reinf:.1f} kg")
    print(f"Slabs Reinforcement: {slab_reinf:.1f} kg")
    print(f"Total Reinforcement: {total_reinf:.1f} kg")
    print(f"Columns Formwork ({form_type}): {col_form:.2f} m²")
    print(f"Beams Formwork ({form_type}): {beam_form_sides + beam_form_bottom:.2f} m²")
    print(f"Slabs Formwork ({form_type}): {slab_form_soffit + slab_form_edges:.2f} m²")
    print(f"Total Formwork: {total_form:.2f} m²")
    
    return {
        'col_conc_m3': col_conc_vol,
        'beam_conc_m3': beam_conc_vol,
        'slab_conc_m3': slab_conc_vol,
        'total_conc_m3': total_conc_vol,
        'col_reinf_kg': col_reinf,
        'beam_reinf_kg': beam_reinf,
        'slab_reinf_kg': slab_reinf,
        'total_reinf_kg': total_reinf,
        'col_form_m2': col_form,
        'beam_form_m2': beam_form_sides + beam_form_bottom,
        'slab_form_m2': slab_form_soffit + slab_form_edges,
        'total_form_m2': total_form
    }

if __name__ == "__main__":
    quantities = calculate_rcc_superstructure()
```

```python
# internal_finishes_takeoff.py
import math

def calculate_internal_finishes():
    print("Internal Finishes Quantity Takeoff Calculator")
    
    # Room inputs
    num_rooms = int(input("Number of rooms: "))
    rooms = []
    for i in range(num_rooms):
        print(f"\nRoom {i+1}:")
        L = float(input("Length (m): "))
        W = float(input("Width (m): "))
        H = float(input("Height (m): "))
        doors = int(input("Number of doors: "))
        door_H = float(input("Door height (m, default 2.1): ") or 2.1)
        door_W = float(input("Door width (m, default 0.9): ") or 0.9)
        windows = int(input("Number of windows: "))
        win_H = float(input("Window height (m, default 1.2): ") or 1.2)
        win_W = float(input("Window width (m, default 1.5): ") or 1.5)
        rooms.append({
            'L': L, 'W': W, 'H': H,
            'doors': doors, 'door_H': door_H, 'door_W': door_W,
            'windows': windows, 'win_H': win_H, 'win_W': win_W
        })
    
    # Materials
    plaster_thick = float(input("\nPlaster thickness (mm, default 15): ") or 15) / 1000
    screed_thick = float(input("Floor screed thickness (mm, default 25): ") or 25) / 1000
    tile_size = float(input("Floor/wall tile size (mm, default 300): ") or 300) / 1000
    skirting_H = float(input("Skirting height (mm, default 100): ") or 100) / 1000
    paint_coats = int(input("Number of paint coats (default 3): ") or 3)
    
    # Calculations
    total_floor_area = 0
    total_wall_area = 0
    total_ceiling_area = 0
    total_door_area = 0
    total_window_area = 0
    total_perim = 0
    
    for r in rooms:
        floor = r['L'] * r['W']
        ceiling = floor
        wall_gross = 2 * (r['L'] + r['W']) * r['H']
        door_area = r['doors'] * r['door_H'] * r['door_W']
        win_area = r['windows'] * r['win_H'] * r['win_W']
        wall_net = wall_gross - door_area - win_area
        perim = 2 * (r['L'] + r['W'])
        
        total_floor_area += floor
        total_wall_area += wall_net
        total_ceiling_area += ceiling
        total_door_area += door_area
        total_window_area += win_area
        total_perim += perim
    
    # Plaster
    wall_plaster = total_wall_area
    ceiling_plaster = total_ceiling_area
    total_plaster_area = wall_plaster + ceiling_plaster
    plaster_vol = total_plaster_area * plaster_thick
    
    # Screed
    screed_area = total_floor_area
    screed_vol = screed_area * screed_thick
    
    # Tiles
    tile_area_floor = screed_area
    tile_area_wall = total_wall_area  # Assume full wall tiling
    tiles_floor = math.ceil(tile_area_floor / (tile_size ** 2))
    tiles_wall = math.ceil(tile_area_wall / (tile_size ** 2))
    
    # Skirting
    skirting_len = total_perim
    skirting_area = skirting_len * skirting_H
    
    # Painting
    paint_wall = total_wall_area * paint_coats
    paint_ceiling = total_ceiling_area * paint_coats
    paint_doors = total_door_area * 2 * paint_coats  # Both sides
    paint_windows = total_window_area * paint_coats  # Frames approx
    total_paint_area = paint_wall + paint_ceiling + paint_doors + paint_windows
    
    # Outputs
    print("\n--- INTERNAL FINISHES QUANTITIES ---")
    print(f"Wall Plaster (15mm): {wall_plaster:.2f} m²")
    print(f"Ceiling Plaster (15mm): {ceiling_plaster:.2f} m²")
    print(f"Total Plaster Volume: {plaster_vol:.3f} m³")
    print(f"Floor Screed (25mm): {screed_area:.2f} m², Vol: {screed_vol:.3f} m³")
    print(f"Floor Tiles (300x300mm): {tiles_floor} no.")
    print(f"Wall Tiles (300x300mm): {tiles_wall} no.")
    print(f"Skirting (100mm high): {skirting_len:.2f} m")
    print(f"Total Painting ({paint_coats} coats): {total_paint_area:.2f} m²")
    
    return {
        'wall_plaster_m2': wall_plaster,
        'ceiling_plaster_m2': ceiling_plaster,
        'plaster_vol_m3': plaster_vol,
        'screed_m2': screed_area,
        'screed_vol_m3': screed_vol,
        'floor_tiles_no': tiles_floor,
        'wall_tiles_no': tiles_wall,
        'skirting_m': skirting_len,
        'paint_m2': total_paint_area
    }

if __name__ == "__main__":
    quantities = calculate_internal_finishes()
```

```python
# doors_windows_takeoff.py
import math

def calculate_doors_windows_takeoff():
    print("Doors and Windows Quantity Takeoff Calculator")
    
    # Type
    item_type = input("Item type (door/window): ").lower()
    
    # Common inputs
    opening_W = float(input("Opening width (m): "))
    opening_H = float(input("Opening height (m): "))
    wall_thick = float(input("Wall thickness (m, default 0.2): ") or 0.2)
    frame_W = float(input("Frame width (mm, default 100): ") or 100) / 1000
    frame_thick = float(input("Frame thickness (mm, default 50): ") or 50) / 1000
    horn_L = float(input("Horn length (mm, default 150): ") or 150) / 1000
    architrave_W = float(input("Architrave width (mm, default 38): ") or 38) / 1000
    architrave_thick = float(input("Architrave thickness (mm, default 14): ") or 14) / 1000
    quadrant_size = float(input("Quadrant size (mm, default 25): ") or 25) / 1000
    lintel_bearing = float(input("Lintel bearing each side (m, default 0.2): ") or 0.2)
    lintel_H = float(input("Lintel height (m, default 0.2): ") or 0.2)
    lintel_W = wall_thick
    reinf_bar_diam = float(input("Reinf bar diam (mm, default 12): ") or 12) / 1000
    num_reinf_bars = int(input("Number of reinf bars (default 4): ") or 4)
    cover = float(input("Reinf cover (mm, default 25): ") or 25) / 1000
    reinf_extra = float(input("Reinf extra length (m, default 0.05): ") or 0.05)
    form_type = input("Formwork type (default timber): ") or "timber"
    
    if item_type == "door":
        leaf_thick = float(input("Door leaf thickness (mm, default 45): ") or 45) / 1000
        leaf_material = input("Leaf material (default flush with plywood): ") or "flush with plywood"
        leaping_size = float(input("Leaping size (mm, default 10): ") or 10) / 1000
        num_doors = int(input("Number of doors: "))
        # Ironmongery
        num_locks = int(input("Number of locks (default 1): ") or 1)
        num_stoppers = int(input("Number of stoppers (default 1): ") or 1)
        num_bolts = int(input("Number of bolts (default 2): ") or 2)
        num_clamps = int(input("Number of clamps (default 5): ") or 5)
    elif item_type == "window":
        glazing_thick = float(input("Glazing thickness (mm, default 5): ") or 5) / 1000
        num_panes = int(input("Number of panes: "))
        has_mullions = input("Has mullions? (yes/no): ").lower() == "yes"
        if has_mullions:
            mullion_L = float(input("Mullion length (m): "))
            mullion_size = float(input("Mullion size (m): "))
        num_windows = int(input("Number of windows: "))
    else:
        print("Invalid type.")
        return {}
    
    # Frame
    frame_H = opening_H + 2 * horn_L
    frame_total_L = 2 * frame_H + opening_W
    frame_qty = frame_total_L  # m
    
    # Architrave
    architrave_H = opening_H - 2 * (architrave_W / 2)  # Adjust for skirting
    architrave_total_L = 2 * architrave_H + opening_W
    architrave_qty = architrave_total_L  # m
    
    # Quadrant mould
    quadrant_qty = architrave_qty  # Same length
    
    if item_type == "door":
        # Leaf
        leaf_W = opening_W - 2 * frame_thick
        leaf_H = opening_H - frame_thick
        leaf_area = leaf_W * leaf_H
        leaf_qty = leaf_area * num_doors  # m²
        
        # Leapings (edging)
        leaping_perim = 2 * (leaf_W + leaf_H)
        leaping_qty = leaping_perim * leaping_size  # m² approx
        
        # Ironmongery
        ironmongery = {
            'locks': num_locks * num_doors,
            'stoppers': num_stoppers * num_doors,
            'bolts': num_bolts * num_doors,
            'clamps': num_clamps * num_doors
        }
    else:
        # Window glazing
        pane_area = (opening_W / num_panes) * opening_H if num_panes else opening_W * opening_H
        glazing_area = pane_area * num_panes * num_windows
        glazing_vol = glazing_area * glazing_thick
        
        if has_mullions:
            mullion_qty = mullion_L * mullion_size ** 2  # Vol if needed
    
    # Lintel
    lintel_L = opening_W + 2 * lintel_bearing
    lintel_vol = lintel_L * lintel_W * lintel_H
    
    # Reinforcement
    reinf_L_each = lintel_L - 2 * cover + 2 * lintel_bearing + reinf_extra
    reinf_total_L = num_reinf_bars * reinf_L_each
    reinf_qty = reinf_total_L  # m
    
    # Formwork
    form_soffit = lintel_L * lintel_W
    form_sides = 2 * lintel_L * lintel_H
    total_form = form_soffit + form_sides
    
    # Adjustments
    opening_area = opening_W * opening_H
    deduct_wall = opening_area * wall_thick
    deduct_frame = opening_W * frame_W
    
    # Outputs
    print("\n--- QUANTITIES ---")
    print(f"Frame ({frame_W:.3f}x{frame_thick:.3f}m): {frame_qty:.2f} m")
    print(f"Architrave ({architrave_W:.3f}x{architrave_thick:.3f}m): {architrave_qty:.2f} m")
    print(f"Quadrant Mould ({quadrant_size:.3f}m): {quadrant_qty:.2f} m")
    if item_type == "door":
        print(f"Door Leaf ({leaf_thick:.3f}m thick): {leaf_qty:.2f} m²")
        print(f"Leapings ({leaping_size:.3f}m): {leaping_qty:.2f} m²")
        print(f"Ironmongery: Locks {ironmongery['locks']} no., Stoppers {ironmongery['stoppers']} no., Bolts {ironmongery['bolts']} no., Clamps {ironmongery['clamps']} no.")
    else:
        print(f"Glazing ({glazing_thick:.3f}m thick): {glazing_area:.2f} m²")
        if has_mullions:
            print(f"Mullions: {mullion_qty:.3f} m³")
    print(f"Lintel Concrete: {lintel_vol:.3f} m³")
    print(f"Reinforcement ({reinf_bar_diam:.3f}m diam): {reinf_qty:.2f} m")
    print(f"Formwork ({form_type}): {total_form:.2f} m²")
    print(f"Deductions: Wall {deduct_wall:.3f} m³, Frame {deduct_frame:.2f} m²")
    
    return {
        'frame_m': frame_qty,
        'architrave_m': architrave_qty,
        'quadrant_m': quadrant_qty,
        'lintel_m3': lintel_vol,
        'reinf_m': reinf_qty,
        'form_m2': total_form,
        'deduct_wall_m3': deduct_wall,
        'deduct_frame_m2': deduct_frame
    }

if __name__ == "__main__":
    quantities = calculate_doors_windows_takeoff()
```

```python
# api.py
from fastapi import APIRouter, HTTPException
import math
from typing import List, Dict, Any
from models import (
    UndergroundTankInput, UndergroundTankOutput,
    SepticTankInput, SepticTankOutput,
    RetainingWallInput, RetainingWallOutput,
    SwimmingPoolInput, SwimmingPoolOutput,
    BasementInput, BasementOutput,
    RCCSuperstructureInput, RCCSuperstructureOutput,
    ElevatedTankInput, ElevatedTankOutput,
    InternalFinishesInput, InternalFinishesOutput,
    DoorsWindowsInput, DoorsWindowsOutput
)

router = APIRouter()

# === UNDERGROUND TANK ===
@router.post("/underground-tank", response_model=UndergroundTankOutput)
def underground_tank(data: UndergroundTankInput):
    try:
        if data.tank_type == "circular":
            if not data.int_diam:
                raise ValueError("int_diam required for circular tank")
            int_radius = data.int_diam / 2
            int_area = math.pi * int_radius ** 2
            ext_diam = data.int_diam + 2*(data.plaster_thick + data.wall_thick + data.mastic_thick + data.brick_thick)
            ext_radius = ext_diam / 2
            ext_area = math.pi * ext_radius ** 2
            excav_diam = ext_diam + 2*data.working_space
            excav_radius = excav_diam / 2
            excav_area = math.pi * excav_radius ** 2
            wall_girth = math.pi * data.int_diam
            form_perim = math.pi * data.int_diam * data.slab_thick
        else:
            int_area = data.int_L * data.int_W
            ext_L = data.int_L + 2*(data.plaster_thick + data.wall_thick + data.mastic_thick + data.brick_thick)
            ext_W = data.int_W + 2*(data.plaster_thick + data.wall_thick + data.mastic_thick + data.brick_thick)
            ext_area = ext_L * ext_W
            excav_L = ext_L + 2*data.working_space
            excav_W = ext_W + 2*data.working_space
            excav_area = excav_L * excav_W
            wall_girth = 2 * (data.int_L + data.int_W)
            form_perim = 2 * (data.int_L + data.int_W) * data.slab_thick
        
        excav_depth = data.depth_int + data.bed_thick + data.blinding_thick + data.cover_to_slab + data.veg_soil
        net_excav_depth = excav_depth - data.veg_soil
        excavation = excav_area * net_excav_depth
        topsoil = excav_area * data.veg_soil
        blinding = excav_area
        concrete_bed = ext_area * data.bed_thick
        wall_vol = wall_girth * data.depth_int * data.wall_thick
        openings = data.num_covers * data.cover_L * data.cover_W
        slab_net = ext_area - openings
        slab_conc = slab_net * data.slab_thick
        backfill_vol = excavation - (concrete_bed + wall_vol + slab_conc + data.blinding_thick * excav_area)
        formwork_soffit = int_area
        formwork_edges = data.num_covers * 2 * (data.cover_L + data.cover_W) * data.slab_thick
        formwork = formwork_soffit + formwork_edges + form_perim
        
        return UndergroundTankOutput(
            topsoil_m3=topsoil,
            excavation_m3=excavation,
            blinding_m2=blinding,
            bed_m3=concrete_bed,
            walls_m3=wall_vol,
            slab_m3=slab_conc,
            formwork_m2=formwork,
            backfill_m3=backfill_vol
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === SEPTIC TANK ===
@router.post("/septic-tank", response_model=SepticTankOutput)
def septic_tank(data: SepticTankInput):
    try:
        int_L = data.tank_int_L
        int_W = data.tank_int_W
        depth_int = data.tank_depth_int
        man_L = data.manhole_int_L
        man_W = data.manhole_int_W
        man_depth = data.manhole_depth_int
        baffles = data.num_baffles
        baffle_L = data.baffle_L
        baffle_thick = data.baffle_thick
        covers = data.num_covers
        cover_L = data.cover_L
        cover_W = data.cover_W
        
        ext_L = int_L + 2*data.wall_thick
        ext_W = int_W + 2*data.wall_thick
        ext_area = ext_L * ext_W
        excav_L = ext_L + 2*data.working_space
        excav_W = ext_W + 2*data.working_space
        excav_area = excav_L * excav_W
        
        clearance = excav_area
        topsoil = excav_area * data.veg_soil_depth
        excavation = excav_area * (depth_int + data.bed_thick_tank + data.blinding_thick + data.cover_soil_depth - data.veg_soil_depth)
        blinding = excav_area
        bed = ext_area * data.bed_thick_tank
        walls = 2*(ext_L + ext_W) * depth_int * data.wall_thick
        slab_openings = covers * cover_L * cover_W
        slab = (ext_area - slab_openings) * data.slab_thick
        form_soffit = int_L * int_W
        form_edges = covers * 2 * (cover_L + cover_W) * data.slab_thick
        form_perim = 2 * (int_L + int_W) * data.slab_thick
        formwork = form_soffit + form_edges + form_perim
        backfill = excavation - (bed + walls + slab + data.blinding_thick * blinding)
        pipes = data.inlet_pipe_L + data.outlet_pipe_L
        
        return SepticTankOutput(
            clearance_m2=clearance,
            veg_soil_m3=topsoil,
            excavation_m3=excavation,
            blinding_m2=blinding,
            bed_m3=bed,
            walls_m3=walls,
            slab_m3=slab,
            formwork_m2=formwork,
            backfill_m3=backfill,
            covers_no=covers,
            pipes_m=pipes
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === RETAINING WALL ===
@router.post("/retaining-wall", response_model=RetainingWallOutput)
def retaining_wall(data: RetainingWallInput):
    try:
        total_height = data.wall_height_below_gl + data.wall_height_above_gl
        avg_width = (data.wall_base_width + data.wall_top_width) / 2
        excav_width = data.found_width + 2 * data.working_space
        excav_area = data.wall_length * excav_width
        veg_soil = excav_area * data.veg_soil_depth
        excavation = excav_area * (data.found_thick + data.wall_height_below_gl)
        disposal = excavation
        blinding = data.wall_length * data.found_width
        found_conc = data.wall_length * data.found_width * data.found_thick
        wall_conc = data.wall_length * avg_width * total_height
        pier_conc = 0
        if data.has_piers == "yes":
            num_piers = math.floor(data.wall_length / data.pier_spacing) + 1
            pier_conc = num_piers * data.pier_width * data.pier_length * data.pier_height
        total_conc = found_conc + wall_conc + pier_conc
        reinf = total_conc * data.reinf_density if data.reinf_incl == "yes" else 0
        form_wall = data.wall_length * total_height * 2
        form_found = data.wall_length * data.found_thick * 2 + data.found_width * data.found_thick * 2
        form_pier = 0
        if data.has_piers == "yes":
            form_pier = num_piers * (2 * data.pier_length * data.pier_height + 2 * data.pier_width * data.pier_height)
        formwork = form_wall + form_found + form_pier
        weepholes = 0
        ballast = 0
        if data.has_weepholes == "yes":
            h_count = math.floor(data.wall_length / data.weephole_spacing) + 1
            v_count = math.floor(total_height / data.weephole_vert_spacing) + 1
            weepholes = h_count * v_count
            if data.has_ballast_pocket == "yes":
                ballast = weepholes * data.pocket_size ** 3
        backfill = data.wall_length * (data.found_width - data.wall_base_width)/2 * (data.found_thick + data.wall_height_below_gl)
        
        return RetainingWallOutput(
            veg_soil_m3=veg_soil,
            excavation_m3=excavation,
            disposal_m3=disposal,
            blinding_m2=blinding,
            found_conc_m3=found_conc,
            wall_conc_m3=wall_conc,
            pier_conc_m3=pier_conc,
            reinf_kg=reinf,
            formwork_m2=formwork,
            weepholes_no=weepholes,
            ballast_pocket_m3=ballast,
            backfill_m3=backfill
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === SWIMMING POOL ===
@router.post("/swimming-pool", response_model=SwimmingPoolOutput)
def swimming_pool(data: SwimmingPoolInput):
    try:
        ext_add = data.wall_thick + data.tanking_thick + data.working_space
        ext_L = data.int_L + 2 * ext_add
        ext_W = data.int_W + 2 * ext_add
        clearance = ext_L * ext_W
        veg_soil = clearance * data.veg_soil_depth
        avg_depth = (data.shallow_depth + data.deep_depth) / 2
        net_excav = avg_depth + data.bed_thick + data.blinding_thick + data.trench_depth - data.veg_soil_depth
        excavation = ext_L * ext_W * net_excav
        trench_perim = 2 * (ext_L + ext_W)
        trench = trench_perim * data.trench_width * data.trench_depth
        found_conc = trench
        blinding = ext_L * ext_W
        slab = ext_L * ext_W * data.bed_thick
        wall_perim = 2 * (data.int_L + data.int_W)
        wall_h = avg_depth + data.wall_height_above
        walls = wall_perim * wall_h * data.wall_thick
        total_conc = found_conc + slab + walls
        reinf = total_conc * data.reinf_density if data.reinf_incl == "yes" else 0
        form_sides = wall_perim * wall_h * 2
        form_soffit = data.int_L * data.int_W
        formwork = form_sides + form_soffit
        backfill = excavation - (found_conc + slab + walls + data.blinding_thick * blinding) if data.backfill_incl == "yes" else 0
        
        return SwimmingPoolOutput(
            clearance_m2=clearance,
            veg_soil_m3=veg_soil,
            excavation_m3=excavation,
            trench_m3=trench,
            found_conc_m3=found_conc,
            blinding_m2=blinding,
            slab_m3=slab,
            walls_m3=walls,
            reinf_kg=reinf,
            formwork_m2=formwork,
            backfill_m3=backfill
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === BASEMENT ===
@router.post("/basement", response_model=BasementOutput)
def basement(data: BasementInput):
    try:
        excav_L = data.ext_L + 2*data.working_space
        excav_W = data.ext_W + 2*data.working_space
        clearance = excav_L * excav_W
        veg_soil = clearance * data.veg_soil_depth
        excavation = clearance * (data.depth_below_gl + data.bed_thick + data.blinding_thick + data.horiz_tanking_thick + data.found_thick - data.veg_soil_depth)
        disposal = excavation + veg_soil if data.disposal_incl == "yes" else 0
        blinding = data.ext_L * data.ext_W
        found_perim = 2 * (data.ext_L + data.ext_W)
        found = found_perim * data.found_L * data.found_thick / 2
        bed = data.ext_L * data.ext_W * data.bed_thick
        horiz_tanking = bed
        rc_wall_perim = 2 * (data.int_L + data.int_W)
        rc_walls = rc_wall_perim * data.depth_below_gl * data.rc_wall_thick
        vert_tanking = rc_wall_perim * data.depth_below_gl
        ext_wall_area = 2 * (data.ext_L + data.ext_W) * data.depth_below_gl
        int_wall_area = 2 * (data.int_L + data.int_W) * data.depth_below_gl
        slab = data.ext_L * data.ext_W * data.slab_thick
        total_conc = found + bed + rc_walls + slab
        reinf = total_conc * data.reinf_density if data.reinf_incl == "yes" else 0
        form_soffit = data.int_L * data.int_W
        form_walls = rc_wall_perim * data.depth_below_gl * 2
        form_slab_edges = 2 * (data.ext_L + data.ext_W) * data.slab_thick
        form_found = found_perim * data.found_thick * 2
        formwork = form_soffit + form_walls + form_slab_edges + form_found
        backfill = excavation - (bed + rc_walls + slab + found + data.blinding_thick * blinding + data.horiz_tanking_thick * horiz_tanking) if data.backfill_incl == "yes" else 0
        
        return BasementOutput(
            clearance_m2=clearance,
            veg_soil_m3=veg_soil,
            excavation_m3=excavation,
            disposal_m3=disposal,
            blinding_m2=blinding,
            found_conc_m3=found,
            bed_m3=bed,
            horiz_tanking_m2=horiz_tanking,
            rc_walls_m3=rc_walls,
            vert_tanking_m2=vert_tanking,
            ext_wall_m2=ext_wall_area,
            int_wall_m2=int_wall_area,
            slab_m3=slab,
            reinf_kg=reinf,
            formwork_m2=formwork,
            backfill_m3=backfill
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === RCC SUPERSTRUCTURE ===
@router.post("/rcc-superstructure", response_model=RCCSuperstructureOutput)
def rcc_superstructure(data: RCCSuperstructureInput):
    try:
        col_conc = sum(c['w'] * c['d'] * c['h'] for c in data.columns)
        beam_conc = sum(b['l'] * b['w'] * b['d'] for b in data.beams)
        slab_conc = sum(s['a'] * s['t'] for s in data.slabs)
        total_conc = col_conc + beam_conc + slab_conc
        col_reinf = col_conc * data.reinf_density
        beam_reinf = beam_conc * data.reinf_density
        slab_reinf = slab_conc * data.reinf_density
        total_reinf = col_reinf + beam_reinf + slab_reinf
        col_form = sum(2 * (c['w'] + c['d']) * c['h'] for c in data.columns)
        beam_form_sides = sum(2 * b['l'] * b['d'] for b in data.beams)
        beam_form_bottom = sum(b['l'] * b['w'] for b in data.beams)
        slab_form_soffit = sum(s['a'] for s in data.slabs)
        slab_form_edges = sum(2 * (math.sqrt(s['a']) * 4) * s['t'] for s in data.slabs)
        total_form = col_form + beam_form_sides + beam_form_bottom + slab_form_soffit + slab_form_edges
        
        return RCCSuperstructureOutput(
            col_conc_m3=col_conc,
            beam_conc_m3=beam_conc,
            slab_conc_m3=slab_conc,
            total_conc_m3=total_conc,
            col_reinf_kg=col_reinf,
            beam_reinf_kg=beam_reinf,
            slab_reinf_kg=slab_reinf,
            total_reinf_kg=total_reinf,
            col_form_m2=col_form,
            beam_form_m2=beam_form_sides + beam_form_bottom,
            slab_form_m2=slab_form_soffit + slab_form_edges,
            total_form_m2=total_form
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === ELEVATED TANK ===
@router.post("/elevated-tank", response_model=ElevatedTankOutput)
def elevated_tank(data: ElevatedTankInput):
    try:
        if data.tank_type == "rectangular":
            int_area = data.tank_int_L * data.tank_int_W
            perim = 2 * (data.tank_int_L + data.tank_int_W)
            ext_L = data.tank_int_L + 2 * data.wall_thick
            ext_W = data.tank_int_W + 2 * data.wall_thick
            ext_area = ext_L * ext_W
        else:
            int_radius = data.tank_diam / 2
            int_area = math.pi * int_radius ** 2
            perim = math.pi * data.tank_diam
            ext_L = data.tank_diam + 2 * data.wall_thick
            ext_area = math.pi * (ext_L / 2) ** 2
        
        found_side = data.base_width + 2 * data.found_extend
        excav_side = found_side + 2 * data.working_space
        excav_area = excav_side ** 2
        veg_soil = excav_area * data.veg_soil
        excavation = excav_area * (data.found_thick)
        blinding = found_side ** 2
        found = found_side ** 2 * data.found_thick
        legs = data.num_legs * data.leg_size ** 2 * data.tower_height
        walls = perim * data.tank_height * data.wall_thick
        slab = ext_area * data.slab_thick
        total_conc = found + legs + walls + slab
        reinf = total_conc * data.reinf_density
        form_found = 4 * found_side * data.found_thick
        form_legs = data.num_legs * 4 * data.leg_size * data.tower_height
        form_walls = perim * data.tank_height * 2
        form_slab = int_area
        form_edges = perim * data.slab_thick
        formwork = form_found + form_legs + form_walls + form_slab + form_edges if data.form_incl == "yes" else 0
        
        return ElevatedTankOutput(
            veg_soil_m3=veg_soil,
            excavation_m3=excavation,
            blinding_m2=blinding,
            found_m3=found,
            legs_m3=legs,
            walls_m3=walls,
            slab_m3=slab,
            total_conc_m3=total_conc,
            reinf_kg=reinf,
            form_m2=formwork
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === INTERNAL FINISHES ===
@router.post("/internal-finishes", response_model=InternalFinishesOutput)
def internal_finishes(data: InternalFinishesInput):
    try:
        total_floor = 0
        total_wall_gross = 0
        total_door = 0
        total_window = 0
        total_perim = 0
        for r in data.rooms:
            floor = r['L'] * r['W']
            wall_gross = 2 * (r['L'] + r['W']) * r['H']
            door_area = r['doors'] * r['door_H'] * r['door_W']
            win_area = r['windows'] * r['win_H'] * r['win_W']
            total_floor += floor
            total_wall_gross += wall_gross
            total_door += door_area
            total_window += win_area
            total_perim += 2 * (r['L'] + r['W'])
        
        wall_net = total_wall_gross - total_door - total_window
        ceiling = total_floor
        plaster_wall = wall_net
        plaster_ceiling = ceiling
        plaster_vol = (plaster_wall + plaster_ceiling) * data.plaster_thick
        screed = total_floor * data.screed_thick
        tiles_floor = math.ceil(total_floor / (data.tile_size ** 2))
        tiles_wall = math.ceil(wall_net / (data.tile_size ** 2))
        skirting = total_perim * data.skirting_H
        paint = (wall_net + ceiling) * data.paint_coats + total_door * 2 * data.paint_coats + total_window * data.paint_coats
        
        return InternalFinishesOutput(
            wall_plaster_m2=plaster_wall,
            ceiling_plaster_m2=plaster_ceiling,
            plaster_vol_m3=plaster_vol,
            screed_m2=total_floor,
            screed_vol_m3=screed,
            floor_tiles_no=tiles_floor,
            wall_tiles_no=tiles_wall,
            skirting_m=total_perim,
            paint_m2=paint
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

# === DOORS & WINDOWS ===
@router.post("/doors-windows", response_model=DoorsWindowsOutput)
def doors_windows(data: DoorsWindowsInput):
    try:
        frame_H = data.opening_H + 2 * data.horn_L
        frame_total = 2 * frame_H + data.opening_W
        arch_H = data.opening_H
        arch_total = 2 * arch_H + data.opening_W
        quadrant = arch_total
        lintel_L = data.opening_W + 2 * data.lintel_bearing
        lintel_vol = lintel_L * data.wall_thick * data.lintel_H
        reinf_L = lintel_L - 2 * data.cover + 2 * data.reinf_extra
        reinf_total = data.num_reinf_bars * reinf_L
        form_soffit = lintel_L * data.wall_thick
        form_sides = 2 * lintel_L * data.lintel_H
        formwork = form_soffit + form_sides
        deduct_wall = data.opening_W * data.opening_H * data.wall_thick
        deduct_frame = data.opening_W * data.frame_W
        
        return DoorsWindowsOutput(
            frame_m=frame_total,
            architrave_m=arch_total,
            quadrant_m=quadrant,
            lintel_m3=lintel_vol,
            reinf_m=reinf_total,
            form_m2=formwork,
            deduct_wall_m3=deduct_wall,
            deduct_frame_m2=deduct_frame
        )
    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))
```

```python
# models.py
from pydantic import BaseModel
from typing import Optional, List, Dict

class RoadworksInput(BaseModel):
    site_length: float
    site_width: float
    house_length: float
    house_width: float
    parking_length: float
    parking_width: float
    driveway_length: float
    driveway_width: float
    bellmouth_r1: float
    bellmouth_r2: float
    veg_depth: float
    excav_depth: float
    murram_depth_avg: float
    hardcore_thick: float
    bitumen_thick: float
    topsoil_thick: float
    kerb_length: float
    fence_length1: float
    fence_length2: float
    num_trees_small: int
    num_trees_large: int
    num_stumps: int
    building_vol: float
    pipe_length: float
    num_mahogany: int
    hedge_length: float
    num_gates: int
    backing_allow: float

class RoadworksOutput(BaseModel):
    clear_m2: float
    trees_small_nos: int
    trees_large_nos: int
    stumps_nos: int
    build_m3: float
    pipe_m: float
    veg_exc_m3: float
    mass_exc_m3: float
    disp_veg_m3: float
    disp_other_m3: float
    prep_exc_m2: float
    murram_m3: float
    hardcore_m2: float
    topsoil_m3: float
    trim_ts_m2: float
    prep_fill_m2: float
    bitumen_m2: float
    kerb_straight_m: float
    kerb_curved_m: float
    grass_m2: float
    mahogany_nos: int
    hedge_m: float
    fence1_m: float
    fence2_m: float
    gates_nos: int

class SubstructureInput(BaseModel):
    plan_type: str
    ext_length: float
    ext_width: float
    has_internal_walls: bool
    int_wall_len: Optional[float] = 0.0
    has_columns: bool
    num_columns: Optional[int] = 0
    col_size: Optional[float] = 0.0
    col_base_size: Optional[float] = 0.0
    col_excav_depth: Optional[float] = 0.0
    has_recess: bool
    recess_type: Optional[str] = ''
    recess_len: Optional[float] = 0.0
    recess_wid: Optional[float] = 0.0
    has_cavity_wall: bool
    cavity_thick: Optional[float] = 0.0
    veg_depth: float
    trench_depth: float
    reduce_level_depth: float
    wall_thick: float
    conc_thick_strip: float
    hardcore_thick: float
    blinding_thick: float
    dpm_thick: float
    anti_termite: bool
    has_formwork: bool
    has_reinforce: bool
    rebar_len: Optional[float] = 0.0
    clear_extra: float

class SubstructureOutput(BaseModel):
    clear_m2: float
    veg_exc_m3: float
    excav_m3: float
    disp_veg_m3: float
    disp_other_m3: float
    conc_m3: float
    wall_m2: float
    backfill_m3: float
    hardcore_m3: float
    blinding_m2: float
    dpm_m2: float
    anti_termite_m2: float
    formwork_m2: float
    reinforce_m: float
    reinstate_m3: float

class DrainageInput(BaseModel):
    veg_depth: float = 0.15
    has_rock: bool
    rock_start_depth: Optional[float] = 0.0
    has_planking: bool
    has_manholes: bool
    num_mh: Optional[int] = 0
    mh_list: Optional[List[Dict]] = None  # List of manhole dicts
    has_pipes: bool
    pipe_runs: Optional[List[Dict]] = None  # List of pipe run dicts

class DrainageOutput(BaseModel):
    veg_excav_m2: float
    pit_excav_m3: float
    rock_m3: float
    conc_bed_m3: float
    wall_m2: float
    bench_nos: int
    channel_m: float
    slab_m3: float
    form_soffit_m2: float
    form_edge_m: float
    plaster_m2: float
    covers_nos: int
    backfill_m3: float
    trench_m3: float
    bedding_m3: float
    pipe_m: float
    backfill_pipe_m3: float

class RoofInput(BaseModel):
    roof_type: str
    build_len: float
    build_wid: float
    wall_thick: float
    overhang: float
    pitch_deg: float
    truss_space: float
    rafter_space: float
    bearer_space: float
    wp_thick: float
    rafter_thick: float
    tie_thick: float
    strut_thick: float
    binder_thick: float
    ridge_thick: float
    fascia_thick: float
    gutter_wid: float
    tile_lap: float
    downpipe_len: float
    has_valley: bool
    valley_len: Optional[float] = 0.0
    has_covering: bool
    has_eaves_verges: bool
    has_rainwater: bool

class RoofOutput(BaseModel):
    wp_m: float
    princ_rafter_m: float
    common_rafter_m: float
    tie_m: float
    ties_m: float
    struts_m: float
    binder_m: float
    ridge_m: float
    hips_m: float
    valleys_m: float
    cover_m2: float
    ridge_cover_m: float
    fascia_m: float
    verge_m: float
    runner_m: float
    bearers_m: float
    soffit_m2: float
    gutter_m: float
    downpipe_m: float

class SuperstructureInput(BaseModel):
    plan_type: str
    ext_len: float
    ext_wid: float
    wall_height: float
    wall_thick: float
    has_internal_walls: bool
    int_wall_total_len: Optional[float] = 0.0
    has_openings: bool
    num_doors: Optional[int] = 0
    door_area: Optional[float] = 0.0
    num_windows: Optional[int] = 0
    win_area: Optional[float] = 0.0
    has_columns: bool
    num_cols: Optional[int] = 0
    col_size: Optional[float] = 0.0
    col_height: Optional[float] = 0.0
    has_form_col: Optional[bool] = False
    has_beams: bool
    beam_len_total: Optional[float] = 0.0
    beam_wid: Optional[float] = 0.0
    beam_depth: Optional[float] = 0.0
    has_form_beam: Optional[bool] = False
    has_slab: bool
    slab_area: Optional[float] = 0.0
    slab_thick: Optional[float] = 0.0
    has_form_slab: Optional[bool] = False
    has_reinf_slab: Optional[bool] = False
    rebar_slab_kg: Optional[float] = 0.0
    has_parapet: bool
    parapet_girth: Optional[float] = 0.0
    parapet_height: Optional[float] = 0.0
    parapet_thick: Optional[float] = 0.0
    has_coping: Optional[bool] = False
    coping_wid: Optional[float] = 0.0
    coping_thick: Optional[float] = 0.0
    has_rainwater: bool
    downpipe_len: Optional[float] = 0.0
    num_downpipes: Optional[int] = 0
    outlet_diam: Optional[float] = 0.0
    has_shoe: Optional[bool] = False
    shoe_len: Optional[float] = 0.0

class SuperstructureOutput(BaseModel):
    wall_m2: float
    col_conc_m3: float
    col_form_m2: float
    beam_conc_m3: float
    beam_form_m2: float
    slab_conc_m3: float
    slab_form_m2: float
    slab_rebar_kg: float
    parapet_m2: float
    coping_m3: float
    downpipe_m: float
    outlets_nos: int
    shoes_nos: int
```

```python
# api.py
from fastapi import APIRouter
from models import *
import math

router = APIRouter()

@router.post("/roadworks", response_model=RoadworksOutput)
def roadworks_takeoff(data: RoadworksInput):
    # Logic from roadworks_takeoff_professional.py
    clear_area = (data.site_length * data.site_width) - (data.house_length * data.house_width)
    tree_g1 = data.num_trees_small
    tree_g2 = data.num_trees_large
    stumps = data.num_stumps
    build_demol = data.building_vol
    pipe_rem = data.pipe_length
    veg_area = clear_area
    veg_vol = veg_area * data.veg_depth
    park_l_adj = data.parking_length + 2 * data.backing_allow
    park_w_adj = data.parking_width + 2 * data.backing_allow
    park_exc_vol = park_l_adj * park_w_adj * data.excav_depth
    drive_l_adj = data.driveway_length + 2 * data.backing_allow
    drive_w_adj = data.driveway_width + 2 * data.backing_allow
    drive_exc_vol = drive_l_adj * drive_w_adj * data.excav_depth
    bell_area1 = (3/14) * math.pi * data.bellmouth_r1**2
    bell_area2 = (3/14) * math.pi * data.bellmouth_r2**2
    bell_area = bell_area1 + bell_area2
    bell_exc_vol = bell_area * data.excav_depth
    total_excav = park_exc_vol + drive_exc_vol + bell_exc_vol
    disposal_veg = veg_vol
    disposal_other = total_excav
    prep_area = (park_l_adj * park_w_adj) + (drive_l_adj * drive_w_adj) + bell_area
    park_area = park_l_adj * park_w_adj
    drive_area = drive_l_adj * drive_w_adj
    murram_vol_park = park_area * data.murram_depth_avg
    murram_vol_drive = drive_area * data.murram_depth_avg
    murram_vol_bell = bell_area * data.murram_depth_avg
    total_murram = murram_vol_park + murram_vol_drive + murram_vol_bell
    hardcore_area = park_area + drive_area + bell_area
    hardcore_vol = hardcore_area * data.hardcore_thick
    grass_area = data.site_width * (data.site_length - data.driveway_width - data.parking_width)
    topsoil_vol = grass_area * data.topsoil_thick
    trim_topsoil = grass_area
    prep_filled_other = hardcore_area
    bitumen_area = park_area + drive_area + bell_area
    bitumen_vol = bitumen_area * data.bitumen_thick
    kerb_straight = data.kerb_length
    kerb_curved = (math.pi * (data.bellmouth_r1 + data.bellmouth_r2)) / 4
    mahogany_nos = data.num_mahogany
    hedge_len_net = data.hedge_length - (data.backing_allow * 2)
    grass_seed_area = grass_area
    fence1_len = data.fence_length1
    fence2_len = data.fence_length2
    gates_nos = data.num_gates
    return RoadworksOutput(
        clear_m2=clear_area, trees_small_nos=tree_g1, trees_large_nos=tree_g2, stumps_nos=stumps,
        build_m3=build_demol, pipe_m=pipe_rem, veg_exc_m3=veg_vol, mass_exc_m3=total_excav,
        disp_veg_m3=disposal_veg, disp_other_m3=disposal_other, prep_exc_m2=prep_area,
        murram_m3=total_murram, hardcore_m2=hardcore_area, topsoil_m3=topsoil_vol,
        trim_ts_m2=trim_topsoil, prep_fill_m2=prep_filled_other, bitumen_m2=bitumen_area,
        kerb_straight_m=kerb_straight, kerb_curved_m=kerb_curved, grass_m2=grass_seed_area,
        mahogany_nos=mahogany_nos, hedge_m=hedge_len_net, fence1_m=fence1_len,
        fence2_m=fence2_len, gates_nos=gates_nos
    )

@router.post("/substructure", response_model=SubstructureOutput)
def substructure_takeoff(data: SubstructureInput):
    # Logic from substructure_takeoff_app.py
    site_clear_area = (data.ext_length + 2*data.clear_extra) * (data.ext_width + 2*data.clear_extra)
    veg_vol = site_clear_area * data.veg_depth
    center_line_wall = 2*(data.ext_length + data.ext_width) + data.int_wall_len - (4*data.wall_thick if data.plan_type == 'rectangle' else 0)
    trench_excav_vol = center_line_wall * data.wall_thick * data.trench_depth
    reduce_level_vol = site_clear_area * data.reduce_level_depth
    col_excav_vol = data.num_columns * (data.col_base_size**2) * data.col_excav_depth if data.has_columns else 0
    total_excav = trench_excav_vol + reduce_level_vol + col_excav_vol
    disposal_veg = veg_vol
    disposal_other = total_excav - veg_vol
    conc_strip_vol = center_line_wall * data.wall_thick * data.conc_thick_strip
    conc_col_vol = data.num_columns * (data.col_base_size**2) * data.conc_thick_strip if data.has_columns else 0
    total_conc = conc_strip_vol + conc_col_vol
    wall_height = data.trench_depth - data.conc_thick_strip
    wall_area = center_line_wall * wall_height
    backfill_soil_vol = (total_excav - total_conc - wall_area * data.wall_thick) * 0.5
    hardcore_vol = site_clear_area * data.hardcore_thick
    blinding_area = site_clear_area
    dpm_area = blinding_area
    anti_termite_area = blinding_area if data.anti_termite else 0
    formwork_area = 0
    if data.has_formwork:
        formwork_perim = 2*(data.ext_length + data.ext_width)
        formwork_area = formwork_perim * data.conc_thick_strip * 2
    reinforce_len = data.rebar_len if data.has_reinforce else 0
    reinstate_perim = 2*(data.ext_length + data.ext_width + (data.col_base_size * data.num_columns if data.has_columns else 0))
    reinstate_vol = (reinstate_perim / 2) * 0.4 * data.veg_depth
    return SubstructureOutput(
        clear_m2=site_clear_area, veg_exc_m3=veg_vol, excav_m3=total_excav,
        disp_veg_m3=disposal_veg, disp_other_m3=disposal_other, conc_m3=total_conc,
        wall_m2=wall_area, backfill_m3=backfill_soil_vol, hardcore_m3=hardcore_vol,
        blinding_m2=blinding_area, dpm_m2=dpm_area, anti_termite_m2=anti_termite_area,
        formwork_m2=formwork_area, reinforce_m=reinforce_len, reinstate_m3=reinstate_vol
    )

@router.post("/drainage", response_model=DrainageOutput)
def drainage_takeoff(data: DrainageInput):
    # Simplified logic; expand with full manhole/pipe loops as in drainage_takeoff_app.py
    total_excav_veg_m2 = total_excav_pit_m3 = total_rock_m3 = 0
    total_conc_bed_m3 = total_wall_m2 = total_bench_nos = total_channel_m = 0
    total_slab_m3 = total_form_soffit_m2 = total_form_edge_m = total_plaster_m2 = 0
    total_cover_nos = total_backfill_m3 = 0
    total_trench_m3 = total_bed_m3 = total_pipe_m = total_backfill_pipe_m3 = 0
    # Implement full logic here...
    return DrainageOutput(
        veg_excav_m2=total_excav_veg_m2, pit_excav_m3=total_excav_pit_m3, rock_m3=total_rock_m3,
        conc_bed_m3=total_conc_bed_m3, wall_m2=total_wall_m2, bench_nos=total_bench_nos,
        channel_m=total_channel_m, slab_m3=total_slab_m3, form_soffit_m2=total_form_soffit_m2,
        form_edge_m=total_form_edge_m, plaster_m2=total_plaster_m2, covers_nos=total_cover_nos,
        backfill_m3=total_backfill_m3, trench_m3=total_trench_m3, bedding_m3=total_bed_m3,
        pipe_m=total_pipe_m, backfill_pipe_m3=total_backfill_pipe_m3
    )

@router.post("/roof", response_model=RoofOutput)
def roof_takeoff(data: RoofInput):
    # Logic from roof_takeoff_app.py
    wp_len = data.build_len + 2 * data.wall_thick
    wp_m = wp_len * 2
    num_truss = math.floor(data.build_len / data.truss_space) + 1
    half_span = (data.build_wid + 2 * data.wall_thick + 2 * data.overhang - 2 * data.fascia_thick - data.gutter_wid) / 2
    rafter_len = half_span / math.cos(math.radians(data.pitch_deg))
    princ_rafter_m = rafter_len * 2 * num_truss
    num_common = math.floor(data.build_len / data.rafter_space) + 1 - num_truss
    common_rafter_m = rafter_len * 2 * num_common
    tie_len = data.build_wid + 2 * data.wall_thick + data.rafter_thick
    tie_m = tie_len * num_truss
    c = half_span * math.cos(math.radians(data.pitch_deg))
    half_c = c / 2
    tie_a = half_c * math.tan(math.radians(data.pitch_deg))
    strut_b = half_c / math.cos(math.radians(data.pitch_deg))
    ties_m = tie_a * 2 * num_truss
    struts_m = strut_b * 2 * num_truss
    binder_len = data.build_len + 2 * data.wall_thick + 2 * data.overhang - 2 * data.fascia_thick - data.gutter_wid
    binder_m = binder_len * 2
    ridge_m = binder_len
    hips_m = 0
    valleys_m = data.valley_len if data.has_valley else 0
    cover_m2 = fascia_m = verge_m = runner_m = bearers_m = soffit_m2 = gutter_m = downpipe_m = 0
    # Add covering, eaves, rainwater logic if enabled...
    return RoofOutput(
        wp_m=wp_m, princ_rafter_m=princ_rafter_m, common_rafter_m=common_rafter_m,
        tie_m=tie_m, ties_m=ties_m, struts_m=struts_m, binder_m=binder_m,
        ridge_m=ridge_m, hips_m=hips_m, valleys_m=valleys_m, cover_m2=cover_m2,
        ridge_cover_m=0, fascia_m=fascia_m, verge_m=verge_m, runner_m=runner_m,
        bearers_m=bearers_m, soffit_m2=soffit_m2, gutter_m=gutter_m, downpipe_m=downpipe_m
    )

@router.post("/superstructure", response_model=SuperstructureOutput)
def superstructure_takeoff(data: SuperstructureInput):
    # Logic from superstructure_takeoff_app.py
    ext_perim = 2 * (data.ext_len + data.ext_wid)
    centerline_wall = ext_perim - 4 * data.wall_thick + data.int_wall_total_len
    wall_area_gross = centerline_wall * data.wall_height
    openings_area = (data.num_doors * data.door_area) + (data.num_windows * data.win_area)
    wall_area_net = wall_area_gross - openings_area
    col_conc_vol = data.num_cols * (data.col_size ** 2) * data.col_height
    form_col_m2 = data.num_cols * 4 * data.col_size * data.col_height if data.has_form_col else 0
    beam_conc_vol = data.beam_len_total * data.beam_wid * data.beam_depth
    form_beam_m2 = data.beam_len_total * (2 * data.beam_depth + data.beam_wid) if data.has_form_beam else 0
    slab_conc_vol = data.slab_area * data.slab_thick
    form_slab_m2 = data.slab_area if data.has_form_slab else 0
    rebar_slab_total = slab_conc_vol * data.rebar_slab_kg
    parapet_area = data.parapet_girth * data.parapet_height
    coping_vol = data.parapet_girth * data.coping_wid * data.coping_thick if data.has_coping else 0
    downpipe_m = data.downpipe_len * data.num_downpipes
    outlets_nos = data.num_downpipes
    shoes_nos = data.num_downpipes if data.has_shoe else 0
    return SuperstructureOutput(
        wall_m2=wall_area_net, col_conc_m3=col_conc_vol, col_form_m2=form_col_m2,
        beam_conc_m3=beam_conc_vol, beam_form_m2=form_beam_m2, slab_conc_m3=slab_conc_vol,
        slab_form_m2=form_slab_m2, slab_rebar_kg=rebar_slab_total, parapet_m2=parapet_area,
        coping_m3=coping_vol, downpipe_m=downpipe_m, outlets_nos=outlets_nos, shoes_nos=shoes_nos
    )
```
